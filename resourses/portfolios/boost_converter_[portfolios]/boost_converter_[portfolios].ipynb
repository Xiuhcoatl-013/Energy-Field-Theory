# Boost Converter [Portfolios]

__Example 06004.__ __Boost Converter Design 1__
Design a boost converter that will have an output of $30 \mathrm{~V}$ from a $12 \mathrm{~V}$ source. Design for continuous inductor current and an output ripple voltage of less than one percent. The load is a resistance of $50 \Omega$. Assume ideal components for this design.

   * __Solution__
>
> First, determine the duty ratio from Eq. 06027
>
>$$
D=1-\frac{V_{s}}{V_{o}}=1-\frac{12}{30}=0.6
$$
>
> If the switching frequency is selected at $25 \mathrm{KHz}$ to be above the audio range, then the minimum inductance for continuous current is determined from Eq. 06032.
>
>$$
L_{\min }=\frac{D(1-D)^{2}(R)}{2 f}=\frac{0.6(1-0.6)^{2}(50)}{2(25,000)}=96 \mu \mathrm{H}
$$
>
> To provide a margin to ensure continuous current, let $L=120 \mu \mathrm{H}$. Note that $L$ and $f$ are selected somewhat arbitrarily and that other combinations will also give continuous current. 
>
> Using Eq. 06028 and Eq. 06025
>
>$$
I_{L}=\frac{V_{s}}{(1-D)^{2}(R)}=\frac{12}{(1-0.6)^{2}(50)}=1.5 \mathrm{~A}
$$
>
> $$
\begin{aligned}
\frac{\Delta i_{L}}{2}=\frac{V_{s} D T}{2 L}=\frac{(12)(0.6)}{(2)(120)(10)^{-6}(25,000)}=1.2 \mathrm{~A} \\
\end{aligned}
$$
>
> $$
\begin{aligned}
I_{\max }=1.5+1.2=2.7 \mathrm{~A} \\
I_{\min }=1.5-1.2=0.3 \mathrm{~A}
\end{aligned}
$$
>
> The minimum capacitance required to limit the output ripple voltage to $1 \%$ is determined from Eq. 06035.
>
> $$
C \geq \frac{D}{R\left(\Delta V_{o} / V_{o}\right) f}=\frac{0.6}{(50)(0.01)(25,000)}=48 \mu \mathrm{F}
$$

__Example 06005.__ __Boost Converter Design 2__
A boost converter is required to have an output voltage of $8 \mathrm{~V}$ and supply a load current of $1 \mathrm{~A}$. The input voltage varies from $2.7 \mathrm{~V}$ to $4.2 \mathrm{~V}$. A control circuit adjusts the duty ratio to keep the output voltage constant. Select the switching frequency. Determine a value for the inductor such that the variation in inductor current is no more than $40 \mathrm{\%}$ percent of the average inductor current for all operating conditions. Determine a value of an ideal capacitor such that the output voltage ripple is no more than $2 \mathrm{\%}$. Determine the maximum capacitor equivalent series resistance for a $2 \mathrm{\%}$ ripple.

   * __Solution__
> Somewhat arbitrarily, choose $200 \mathrm{kHz}$ for the switching frequency. The circuit must be analyzed for both input voltage extremes to determine the worst-case condition. For $V_{s}=2.7 \mathrm{~V}$ the duty ratio is determined from Eq. 06027.
>
> $$
D=1-\frac{V_{s}}{V_{o}}=1-\frac{2.7}{8}=0.663
$$
>
> Average inductor current is determined from Eq. 06028.
>
> $$
I_{L}=\frac{V_{o} I_{o}}{V_{s}}=\frac{8(1)}{2.7}=2.96 \mathrm{~A}
$$
>
> The variation in inductor current to meet the 40 percent specification is then $\Delta i_{L}=0.4(2.96)=1.19 \mathrm{~A}$. The inductance is then determined from Eq. 06033
>
> $$
L=\frac{V_{s} D}{\Delta i_{L} f}=\frac{2.7(0.663)}{1.19(200,000)}=7.5 \mu \mathrm{H}
$$
>
> Repeating the calculations for $V_{s}=4.2 \mathrm{~V}$
>
>$$
D=1-\frac{V_{s}}{V_{0}}=1-\frac{4.2}{8}=0.475
$$
>
>$$
I_{L}=\frac{V_{o} I_{o}}{V_{s}}=\frac{8(1)}{4.2}=1.90 \mathrm{~A}
$$
>
>The variation in inductor current for this case is $\Delta i_{L}=0.4(1.90)=0.762 \mathrm{~A},$ and
>
>$$
L=\frac{V_{s} D}{\Delta i_{L} f}=\frac{4.2(0.475)}{0.762(200,000)}=13.1 \mu \mathrm{H}
$$
>
>The inductor must be $13.1 \mu \mathrm{H}$ to satisfy the specifications for the total range of input voltages.
>
>Equation 06035, using the maximum value of $D$, gives the minimum capacitance as
>
>$$
C=\frac{D}{R\left(\Delta V_{o} / V_{o}\right) f}=\frac{D}{\left(V_{o} / I_{o}\right)\left(\Delta V_{o} / V_{o}\right) f}=\frac{0.663}{(8 / 1)(0.02)(200,000)}=20.7 \mu \mathrm{F}
$$
>
>The maximum ESR is determined from Eq. 06036, using the maximum peak-to-peak variation in capacitor current. The peak-to-peak variation in capacitor current is the same as maximum inductor current. The average inductor current varies from $2.96 \mathrm{~A}$ at $V_{s}=2.7 \mathrm{~V}$ to $1.90 \mathrm{~A}$ at $V_{s}=4.2 \mathrm{~V}$. The variation in inductor current is $0.762 \mathrm{~A}$ for $V_{s}=4.2 \mathrm{~A},$ but it must be recalculated for $V_{s}=2.7 \mathrm{~V}$ using the $13.1 \mu \mathrm{H}$ value selected, yielding
>
>$$
\Delta i_{L}=\frac{V_{s} D}{L f}=\frac{2.7(0.663)}{13.1(10)^{-6}(200,000)}=0.683 \mathrm{~A}
$$
>
>Maximum inductor current for each case is then computed as
>
>$$
\begin{array}{l}
I_{L, \max , 2.7 \mathrm{~V}}=I_{L}+\frac{\Delta i_{L}}{2}=2.96+\frac{0.683}{2}=3.30 \mathrm{~A} \\
I_{L, \max , 4.2 \mathrm{~V}}=I_{L}+\frac{\Delta i_{L}}{2}=1.90+\frac{0.762}{2}=2.28 \mathrm{~A}
\end{array}
$$
>
>This shows that the largest peak-to-peak current variation in the capacitor will be $3.30 \mathrm{~A}$. The output voltage ripple due to the capacitor ESR must be no more than $(0.02)(8)=0.16 \mathrm{~V}$. Using Eq. 06036
>
>$$
\Delta V_{o, \mathrm{ESR}}=\Delta i_{C} r_{C}=I_{L, \max } r_{C}=3.3 r_{C}=0.16 \mathrm{~V}
$$
>
>which gives
>
>$$
r_{C}=\frac{0.16 \mathrm{~V}}{3.3 \mathrm{~A}}=48 \mathrm{~m} \Omega
$$
>
>In practice, a capacitor that has an ESR of $48 \mathrm{~m} \Omega$ or less could have a capacitance value much larger than the $20.7 \mu \mathrm{F}$ calculated.

__Example 06010.__ __Boost Converter with Discontinuous Current__
The boost converter of Fig. 06008a has parameters

$$
\begin{aligned}
V_{s} &=20 \mathrm{V} \\
D &=0.6 \\
L &=100 \mu \mathrm{H} \\
R &=50 \Omega \\
C &=100 \mu \mathrm{F} \\
f &=15 \mathrm{kHz}
\end{aligned}
$$

(a) Verify that the inductor current is discontinuous, (b) determine the output voltage, and (c) determine the maximum inductor current.



   * __Solution__
>
>(a) First assume that the inductor current is continuous and compute the minimum from [Eq. 06030], resulting in $I_{\min }=-1.5 A$. Negative inductor current is not possible, indicating discontinuous current.
(b) The [Eq. 06102] gives the output voltage
> 
> $$
V_{o}=\frac{V_{s}}{2}\left(1+\sqrt{1+\frac{2 D^{2} R}{L f}}\right)=\frac{20}{2}\left[1+\sqrt{1+\frac{2(0.6)^{2}(50)}{100(10)^{-6}(15,000)}}\right] = 60 \mathrm{V}
$$
> 
> Note that a boost converter with the same duty ratio operating with continuous current would have an output of $50 \mathrm{V}$.
(c) The maximum inductor current is determined from [Eq. 06099].
> 
> $$
I_{\max }=\frac{V_{s} D}{L f}=\frac{(20)(0.6)}{100(10)^{-6}(15,000)}=8 \mathrm{A}
$$
