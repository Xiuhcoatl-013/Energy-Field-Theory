# Buck Converter [Portfolios]

[__Example 06001__] __Buck Converter__

The buck dc-dc converter of Fig. 06003a has the following parameters:

$$
\begin{aligned}
V_{s} &=50 \mathrm{V} \\
D &=0.4 \\
L &=400 \mu \mathrm{H} \\
C &=100 \mu \mathrm{F} \\
f &=20 \mathrm{kHz} \\
R &=20 \Omega
\end{aligned}
$$

Assuming ideal components, calculate ( $a$ ) the output voltage $V_{o},$ (b) the maximum and minimum inductor current, and (c) the output voltage ripple.



> __■ Solution__
>
>__(a)__ The inductor current is assumed to be continuous, and the output voltage is computed from [Eq. 06009]
> 
> $$
V_{o}=V_{s} D=(50)(0.4)=20 \mathrm{V}
$$
> 
>__(b)__ Maximum and minimum inductor currents are computed from [Eq. 06011] and [Eq. 06012].
> 
> $$
\begin{aligned}
I_{\max } &=V_{o}\left(\frac{1}{R}+\frac{1-D}{2 L f}\right) \\
=20\left[\frac{1}{20}+\frac{1-0.4}{2(400)(10)^{-6}(20)(10)^{3}}\right] \\
&=1+\frac{1.5}{2}=1.75 \mathrm{A}
\end{aligned}
$$
>
> $$
\begin{aligned}
I_{\min } &=V_{o}\left(\frac{1}{R}-\frac{1-D}{2 L f}\right) \\
&=1-\frac{1.5}{2}=0.25 \mathrm{A}
\end{aligned}
$$
> 
> The average inductor current is $1 \mathrm{A},$ and $\Delta i_{L}=1.5 \mathrm{A}$. Note that the minimum inductor current is positive, verifying that the assumption of continuous current was valid.
>
>__(c)__ The output voltage ripple is computed from [Eq. 06019].
> 
> $$
\begin{aligned}
\frac{\Delta V_{o}}{V_{o}}=\frac{1-D}{8 L C f^{2}} &=\frac{1-0.4}{8(400)(10)^{-6}(100)(10)^{-6}(20,000)^{2}} \\
&=0.00469=0.469 \%
\end{aligned}
$$
> 
> since the output ripple is sufficiently small, the assumption of a constant output voltage was reasonable.

[__Example 06002__] __Buck Converter Design 1__

Design a buck converter to produce an output voltage of $18 \mathrm{V}$ across a $10\Omega$ load resistor. The output voltage ripple must not exceed $0.5 \%$. The de supply is $48 \mathrm{V}$. Design for continuous inductor current. Specify the duty ratio, the switching frequency, the values of the inductor and capacitor, the peak voltage rating of each device, and the rms current in the inductor and capacitor. Assume ideal components.



> __■ Solution__
Using the buck converter circuit in Fig. 06003a, the duty ratio for continuous-current operation is determined from [Eq. 06009]
> 
> $$
D=\frac{V_{o}}{V_{s}}=\frac{18}{48}=0.375
$$
> 
> The switching frequency and inductor size must be selected for continuous-current operation. Let the switching frequency arbitrarily be $40 \mathrm{kHz},$ which is well above the audio range and is low enough to keep switching losses small. The minimum inductor size is determined from [Eq. 06014]
> 
> $$
L_{\min }=\frac{(1-D)(R)}{2 f}=\frac{(1-0.375)(10)}{2(40,000)}=78 \mu \mathrm{H}
$$
> 
> Let the inductor be 25 percent larger than the minimum to ensure that inductor current is continuous.
> 
> $$
L=1.25 L_{\min }=(1.25)(78 \mu \mathrm{H})=97.5 \mu \mathrm{H}
$$
> 
> Average inductor current and the change in current are determined from [Eq. 06010] and [Eq. 06017]
> 
> $$
\begin{array}{c}
I_{L}=\frac{V_{o}}{R}=\frac{18}{10}=1.8 \mathrm{A} \\
\Delta i_{L}=\left(\frac{V_{s}-V_{o}}{L}\right) D T=\frac{48-18}{97.5(10)^{-6}}(0.375)\left(\frac{1}{40,000}\right)=2.88 \mathrm{A}
\end{array}
$$
> 
> The maximum and minimum inductor currents are determined from [Eq. 06011] and [Eq. 06012]
> 
> $$
I_{\max }=I_{L}+\frac{\Delta i_{L}}{2}=1.8+1.44=3.24 \mathrm{A}
$$
> 
> $$
I_{\min }=I_{L}-\frac{\Delta i_{L}}{2}=1.8-1.44=0.36 \mathrm{A}
$$
> 
> The inductor must be rated for rms current, which is computed as in __Chap. 02__ (see Example 06008). For the offset triangular wave,
> 
> $$
I_{L, \mathrm{ms}}=\sqrt{I_{L}^{2}+\left(\frac{\Delta i_{L} / 2}{\sqrt{3}}\right)^{2}}=\sqrt{(1.8)^{2}+\left(\frac{1.44}{\sqrt{3}}\right)^{2}}=1.98 \mathrm{A}
$$
> 
> The capacitor is selected using [Eq. 06020].
> 
> $$
C=\frac{1-D}{8 L\left(\Delta V_{o} / V_{o}\right) f^{2}}=\frac{1-0.375}{8(97.5)(10)^{-6}(0.005)(40,000)^{2}}=100 \mu \mathrm{F}
$$
> 
> Peak capacitor current is $\Delta i_{L} / 2=1.44 \mathrm{A},$ and $\mathrm{rms}$ capacitor current for the triangular waveform is $1.44 / \sqrt{3}=0.83$ A. The maximum voltage across the switch and diode is $V_{s}$, or $48 \mathrm{V}$. The inductor voltage when the switch is closed is $V_{s}-V_{o}=48-18=30 \mathrm{V}$. The inductor voltage when the switch is open is $V_{o}=18 \mathrm{V}$. Therefore, the inductor must withstand $30 \mathrm{V}$. The capacitor must be rated for the $18 \mathrm{V}$ output.

[__Example 06003__] __Buck Converter Design 2__

Power supplies for telecommunications applications may require high currents at low voltages. Design a buck converter that has an input voltage of $3.3 \mathrm{V}$ and an output voltage of $1.2 \mathrm{V}$. The output current varies between $4$ and $6 \mathrm{A}$. The output voltage ripple must not exceed $2 \%$. Specify the inductor value such that the peak-to-peak variation in inductor current does not exceed $40 \%$ of the average value. Determine the required rms current rating of the inductor and of the capacitor. Determine the maximum equivalent series resistance of the capacitor.



> __■ Solution__
Because of the low voltage and high output current in this application, the synchronous rectification buck converter of Fig. 06007 is used. The duty ratio is determined from [Eq. 06009].
> 
> $$
D=\frac{V_{o}}{V_{s}}=\frac{1.2}{3.3}=0.364
$$
> 
> The switching frequency and inductor size must be selected for continuous-current operation. Let the switching frequency arbitrarily be $500 \mathrm{kHz}$ to give a good tradeoff between small component size and low switching losses.
> 
> The average inductor current is the same as the output current. Analyzing the circuit for an output current of $4 \mathrm{A}$,
> 
> $$
\begin{array}{c}
I_{L}=I_{o}=4 \mathrm{A} \\
\Delta i_{L}=(40 \%)(4)=1.6 \mathrm{A}
\end{array}
$$
> 
> Using [Eq. 06016]
> 
> $$
L=\left(\frac{V_{s}-V_{o}}{\Delta i_{L} f}\right) \mathrm{D}=\frac{3.3-1.2}{(1.6)(500,000)}(0.364)=0.955 \mu \mathrm{H}
$$
> 
> Analyzing the circuit for an output current of $6 \mathrm{A}$,
> 
> $$
\begin{array}{c}
I_{L}=I_{o}=6 \mathrm{A} \\
\Delta i_{L}=(40 \%)(6)=2.4 \mathrm{A}
\end{array}
$$
> 
> resulting in
>
> $$
L=\left(\frac{V_{s}-V_{o}}{\Delta i_{L} f}\right) D=\frac{3.3-1.2}{(2.4)(500,000)}(0.364)=0.636 \mu \mathrm{H}
$$
> 
> since $0.636 \mu \mathrm{H}$ would be too small for the $4\mathrm{A}$ output, use $L=0.955 \mu \mathrm{H},$ which would be rounded to $1 \mu \mathrm{H}$ Inductor rms current is determined from
> 
> $$
I_{L, \mathrm{rms}}=\sqrt{I_{L}^{2}+\left(\frac{\Delta i_{L} / 2}{\sqrt{3}}\right)^{2}}
$$
> 
> For more details on the last equations, refer to the section [__Electrical Quantities and Power Computations__](https://github.com/Xiuhcoatl-013/Energy-Field-Theory/blob/master/notes/electrical_quantities_and_power_computations/electrical_quantities_and_power_computations.ipynb). From [Eq. 06015], the variation in inductor current is $1.6 \mathrm{A}$ for each output current. Using the $6\mathrm{A}$ output current, the inductor must be rated for an $\mathrm{rms}$ current of
> 
> $$
I_{L, \mathrm{rms}}=\sqrt{6^{2}+\left(\frac{0.8}{\sqrt{3}}\right)^{2}}=6.02 \mathrm{A}
$$
> 
> Note that the average inductor current would be a good approximation to the rms current since the variation is relatively small. Using $L=1 \mu \mathrm{H}$ in [Eq. 06020], the minimum capacitance is determined as
> 
> $$
C=\frac{1-D}{8 L\left(\Delta V_{o} / V_{o}\right) f^{2}}=\frac{1-0.364}{8(1)(10)^{-6}(0.02)(500,000)^{2}}=0.16 \mu \mathrm{F}
$$
> 
> The allowable output voltage ripple of 2 percent is $(0.02)(1.2)=24 \mathrm{mV} .$ The maximum ESR is computed from [Eq. 06023]
> 
> $$
\Delta V_{o} \approx r_{C} \Delta i_{C}=r_{C} \Delta i_{L}
$$
> 
> or
> 
> $$
r_{C}=\frac{\Delta V_{o}}{\Delta i_{C}}=\frac{24 \mathrm{mV}}{1.6 \mathrm{A}}=15 \mathrm{m} \Omega
$$   [Eq. 060]
> 
> At this point, the designer would search manufacturer 's specifications for a capacitor having $15-\mathrm{m} \Omega$ ESR. The capacitor may have to be much larger than the calculated value of $0.16 \mu \mathrm{F}$ to meet the ESR requirement. Peak capacitor current is $\Delta i_{L} / 2=0.8 \mathrm{A}$ and rms capacitor current for the triangular waveform is $0.8 / \sqrt{3}=0.46 \mathrm{A}$

[__Example 06009]__ __Buck Converter with Discontinuous Current__

For the buck converter of Fig. 06003a

$$
\begin{aligned}
V_{s} &=24 \mathrm{V} \\
L &=200 \mu \mathrm{H} \\
R &=20 \Omega \\
C &=1000 \mu \mathrm{F}
\end{aligned}
$$

$f=10 \mathrm{kHz} \quad$ switching frequency

$$
D=0.4
$$

(a) Show that the inductor current is discontinuous, ( $b$ ) Determine the output voltage $V_{o^{*}}$



> __■ Solution__
>
> __(a)__ For discontinuous current, $D_{1}<1-D$, and $D_{1}$ is calculated from [Eq. 06095].
> 
> $$
\begin{array}{c}
D_{1}=\frac{-D+\sqrt{D^{2}+8 L / R T}}{2} \\
=\frac{1}{2}\left(-0.4+\sqrt{0.4^{2}+\frac{8(200)(10)^{-6}(10,000)}{20}}\right)=0.29
\end{array}
$$
> 
> Comparing $D_{1}$ to $1-D, 0.29<(1-0.4)$ shows that the inductor current is discontinuous. Alternatively, the minimum inductor current computed from [Eq. 06012] is $I_{\min }= -0.96 A$. since negative inductor current is not possible, inductor current must be discontinuous.
> 
> __(b)__ since $D_{1}$ is calculated and discontinuous current is verified, the output voltage can be computed from [Eq. 06096]
> 
> $$
V_{o}=V_{s}\left(\frac{D}{D+D_{1}}\right)=20\left(\frac{0.4}{0.4+0.29}\right)=13.9 \mathrm{V}
$$
> 
> Figure 06021 shows the relationship between output voltage and duty ratio for the buck converter of Example 06009. All parameters except $D$ are those of Example 06009. Note the linear relationship between input and output for continuous current and the nonlinear relationship for discontinuous current. For a given duty ratio, the output voltage is greater for discontinuous-current operation than it would be if current were continuous.
