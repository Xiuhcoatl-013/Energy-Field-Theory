__Example 4. RMS Values of Square Pulse Waveform__
Determine the $\mathrm{rms}$ value of the periodic pulse waveform that has a duty ratio of $D$ as shown in Fig. 6

<center> Figure 6.$\quad$ Pulse waveform for Example 4. </center>
<br><br/>
<br><br/>



> __■ Solution__
> 
> The voltage is expressed as
>
> $$
v(t)=\left\{\begin{array}{lc}
V_{m} & 0<t<D T \\
0 & D T<t<T
\end{array}\right.
$$
>
> Using [Eq. 37] to determine the $\mathrm{rms}$ value of the waveform gives
>
> $$
V_{\mathrm{ms}}=\sqrt{\frac{1}{T} \int_{0}^{T} v^{2}(t) d t}=\sqrt{\frac{1}{T}\left(\int_{0}^{D T} V_{m}^{2} d t+\int_{D T}^{T} 0^{2} d t\right)}=\sqrt{\frac{1}{T}\left(V_{m}^{2} D T\right)}
$$
>
> yielding
>
> $$
V_{\mathrm{rms}}=V_{m} \sqrt{D}
$$

__Example 5. RMS Values of Sinusoids__
Determine the rms values of __(a)__ a sinusoidal voltage of $v(t)=V_{m} \sin (\omega t)$,__(b)__ a full-wave rectified sine wave of $v(t)=\left|V_{m} \sin (\omega t)\right|,$ and __(c)__ a half-wave rectified sine wave of $v(t)=V_{m} \sin (\omega t)$ for $0<t<T / 2$ and zero otherwise.

> __■ Solution__
> 
> __(a)__ The rms value of the sinusoidal voltage is computed from [Eq. 37]:
> 
> $$
V_{\mathrm{ms}}=\sqrt{\frac{1}{T} \int_{0}^{T} V_{m}^{2} \sin ^{2}(\omega t) d t}
$$
> 
> where 
> 
> $$
T=\frac{2 \pi}{\omega}
$$
> 
> An equivalent expression uses $\omega t$ as the variable of integration. Without showing the details of the integration, the result is
> 
> $$
V_{\mathrm{ms}}=\sqrt{\frac{1}{2 \pi} \int_{0}^{2 \pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)}=\frac{V_{m}}{\sqrt{2}}
$$
> 
> Note that the rms value is independent of the frequency.
> 
> __(b)__ [Equation 37] can be applied to the full-wave rectified sinusoid, but the results of part __(a)__ can also be used to advantage. The rms formula uses the integral of the square of the function. The square of the sine wave is identical to the square of the full-wave rectified sine wave, so the $\mathrm{rms}$ values of the two waveforms are identical:
> 
> $$
V_{\mathrm{ms}}=\frac{V_{m}}{\sqrt{2}}
$$
> 
> __(c)__ Equation ( $2-37$ ) can be applied to the half-wave rectified sinusoid.
> 
> $$
\left.V_{\mathrm{mis}}=\sqrt{\frac{1}{2 \pi}\left(\int_{0}^{\pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)+\int_{\pi}^{2 \pi} 0^{2} d(\omega t)\right.}\right)=\sqrt{\frac{1}{2 \pi} \int_{0}^{\pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)}
$$
> 
> The result of part (a) will again be used to evaluate this expression. The square of the function has one-half the area of that of the functions in $(a)$ and $(b)$. That is,
> 
> $$
V_{\mathrm{mns}}=\sqrt{\frac{1}{2 \pi} \int_{0}^{\pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)}=\sqrt{\left(\frac{1}{2}\right) \frac{1}{2 \pi} \int_{0}^{2 \pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)}
$$
> 
> Taking the $1 / 2$ outside of the square root gives
> 
> $$
V_{\mathrm{ms}}=(\sqrt{\frac{1}{2}}) \sqrt{\frac{1}{2 \pi} \int_{0}^{2 \pi} V_{m}^{2} \sin ^{2}(\omega t) d(\omega t)}
$$
>
> The last term on the right is the $\mathrm{rms}$ value of a sine wave which is known to be $V_{m} / \sqrt{2},$ so the rms value of a half-wave rectified sine wave is
>
> $$
V_{\mathrm{ms}}=\sqrt{\frac{1}{2}} \frac{V_{m}}{\sqrt{2}}=\frac{V_{m}}{2}
$$
>

__Example 6. Neutral Conductor Current in a Three-Phase System__
An office complex is supplied from a three-phase four-wire voltage source (Fig. 8a).
The load is highly nonlinear as a result of the rectifiers in the power supplies of the equipment,
and the current in each of the three phases is shown in Fig. 8b. The neutral current is the sum of the phase currents. If the rms current in each phase conductor is known to be $20 \mathrm{A}$, determine the rms current in the neutral conductor.

> __■ Solution__
>
> [Equation 38] may be applied to this case. Noting by inspection that the area of the square of the current function in the neutral $i_{n},$ is 3 times that of each of the phases $i_{a}$ (Fig. 8c)
> 
> $$
I_{n, \mathrm{mns}}=\sqrt{\frac{1}{T} \int_{0}^{T} i_{n}^{2}(t) d(t)}=\sqrt{3\left(\frac{1}{T} \int_{0}^{T} i_{a}^{2}(t) d(t)\right)}=\sqrt{3} I_{a, \mathrm{rms}}
$$
> 
> The rms current in the neutral is therefore
> 
> $$
I_{n, \mathrm{mns}}=\sqrt{3}(20)=34.6 \mathrm{A}
$$
> 
> Note that the rms neutral current is larger than the phase currents for this situation. This is much different from that for balanced linear loads where the line currents are sinusoids which are displaced by $120^{\circ}$ and sum to zero. Three-phase distribution systems supplying highly nonlinear loads should have a neutral conductor capable of carrying $\sqrt{3}$ times as much current as the line conductor.
> 
> Finally note that [Eq. 40] also could be applied in the current example [Example 6] to obtain the rms value of the neutral current.

__Example 7. RMS Value of the Sum of Waveforms__
Determine the effective (rms) value of $v(t)=4+8 \sin \left(\omega_{1} t+10^{\circ}\right)+5 \sin \left(\omega_{2} t+50^{\circ}\right)$ for __(a)__ $\omega_{2}=2 \omega_{1}$ and __(b)__ $\omega_{2}=\omega_{1}$

> __■ Solution__
>
> __(a)__ The rms value of a single sinusoid is $V_{m} / \sqrt{2}$, and the rms value of a constant is the constant. When the sinusoids are of different frequencies, the terms are orthogonal and [Eq. 39] applies.
> 
> $$
V_{\mathrm{rms}}=\sqrt{V_{1, \mathrm{ms}}^{2}+V_{2, \mathrm{rms}}^{2}+V_{3, \mathrm{mns}}^{2}}=\sqrt{4^{2}+\left(\frac{8}{\sqrt{2}}\right)^{2}+\left(\frac{5}{\sqrt{2}}\right)^{2}}=7.78 \mathrm{V}
$$
> 
> __(b)__ For sinusoids of the same frequency, [Eq. 39] does not apply because the integral of the cross product over one period is not zero. First combine the sinusoids using phasor addition:
> 
> $$
8 \angle 10^{\circ}+5 \angle 50^{\circ}=12.3 \angle 25.2^{\circ}
$$
> 
> The voltage function is then expressed as
> 
> $$
v(t)=4+12.3 \sin \left(\omega_{1} t+25.2^{\circ}\right) \mathrm{V}
$$
> 
> The rms value of this voltage is determined from [Eq. 39] as
> 
> $$
V_{\mathrm{rms}}=\sqrt{4^{2}+\left(\frac{12.3}{\sqrt{2}}\right)^{2}}=9.57 \mathrm{V}
$$

__Example 8. RMS Value of Triangular Waveforms__
__(a)__ A triangular current waveform like that shown in [Fig. 9a] is commonly encountered in de power supply circuits. Determine the rms value of this current. 
__(b)__ Determine the rms value of the offset triangular waveform in [Fig. 9b].

> __■ Solution__
> 
> __(a)__ The current is expressed as
> $$
i(t)=\left\{\begin{array}{ll}
\frac{2 I_{m}}{t_{1}} t-I_{m} & 0<t<t_{1} \\
\frac{-2 I_{m}}{T-t_{1}} t+\frac{I_{m}\left(T+t_{1}\right)}{T-t_{1}} & t_{1}<t<T
\end{array}\right.
$$
> 
> The rms value is determined from Eq. $(2-38)$.
> 
> $$
I_{\mathrm{mis}}^{2}=\frac{1}{T}\left[\int_{0}^{t_{1}}\left(\frac{2 I_{m}}{t_{1}} t-I_{m}\right)^{2} d t+\int_{t_{1}}^{T}\left(\frac{-2 I_{m}}{T-t_{1}} t+\frac{I_{m}\left(T+t_{1}\right)}{T-t_{1}}\right)^{2} d t\right]
$$
> 
> <br><br/>
> <p align="center"><img src="images/triangular_waveformn.png" style="width:30%"></p>
> <center> Figure 9.$\quad$ (a) Triangular waveform for Example 8; (b) offset triangular waveform. > </center>
> <br><br/>
> <br><br/>
> 
> The details of the integration are quite long, but the result is simple: The rms value of a triangular current waveform is
>
> $$
I_{\mathrm{mns}}=\frac{I_{m}}{\sqrt{3}}
$$
>
> __(b)__ The rms value of the offset triangular waveform can be determined by using the result of part (a). since the triangular waveform of part (a) contains no de component, the de signal and the triangular waveform are orthogonal, and [Eq. 40] applies.
>
> $$
I_{\mathrm{ms}}=\sqrt{I_{1, \mathrm{ms}}^{2}+I_{2, \mathrm{ms}}^{2}}=\sqrt{\left(\frac{I_{m}}{\sqrt{3}}\right)^{2}+I_{\mathrm{dc}}^{2}}=\sqrt{\left(\frac{2}{\sqrt{3}}\right)^{2}+3^{2}}=3.22 \mathrm{A}
$$

__Example 9. Nonsinusoidal Source and Linear Load__
A nonsinusoidal periodic voltage has a Fourier series of $v(t)=10+20 \cos \left(2 \pi 60 t-25^{\circ}\right)$ $+30 \cos \left(4 \pi 60 t+20^{\circ}\right) \mathrm{V} .$ This voltage is connected to a load that is a $5 \Omega$ resistor and a $15 \mathrm{mH}$ inductor connected in series as in Fig. 11. Determine the power absorbed by the load.

<br><br/>
<p align="center"><img src="images/circuit_example_9.png" style="width:25%"></p>
<center> Figure 11.$\quad$ Circuit for Example 9. </center>
<br><br/>
<br><br/>

> __■ Solution__ 
> 
> Current at each source frequency is computed separately. The de current term is
> 
> $$
I_{0}=\frac{V_{0}}{R}=\frac{10}{5}=2 \mathrm{A}
$$
> 
> The amplitudes of the ac current terms are computed from phasor analysis:
> 
> $$
\begin{array}{l}
I_{1}=\frac{V_{1}}{R+j \omega_{1} L}=\frac{20 \angle\left(-25^{\circ}\right)}{5+j(2 \pi 60)(0.015)}=2.65 \angle\left(-73.5^{\circ}\right) \mathrm{A} \\
I_{2}=\frac{V_{2}}{R+j \omega_{2} L}=\frac{30 \angle 20^{\circ}}{5+j(4 \pi 60)(0.015)}=2.43 \angle\left(-46.2^{\circ}\right) \mathrm{A}
\end{array}
$$
> 
> Load current can then be expressed as
> 
> $$
i(t)=2+2.65 \cos \left(2 \pi 60 t-73.5^{\circ}\right)+2.43 \cos \left(4 \pi 60 t-46.2^{\circ}\right) \mathrm{A}
$$
> 
> Power at each frequency in the Fourier series is determined from [Eq. 59]:
> 
> $$
\begin{array}{ll}
\text { dc term: } & P_{0}=(10 \mathrm{V})(2 \mathrm{A})=20 \mathrm{W} \\
\omega=2 \pi 60: & P_{1}=\frac{(20)(2.65)}{2} \cos \left(-25^{\circ}+73.5^{\circ}\right)=17.4 \mathrm{W} \\
\omega=4 \pi 60: & P_{2}=\frac{(30)(2.43)}{2} \cos \left(20^{\circ}+46^{\circ}\right)=14.8 \mathrm{W}
\end{array}
$$
> 
> Total power is then
> 
> $$
P=20+17.4+14.8=52.2 \mathrm{W}
$$
> 
> Power absorbed by the load can also be computed from $I_{\mathrm{rms}}^{2} R$ in this circuit because the average power in the inductor is zero.
> 
> $$
P=I_{\mathrm{mns}}^{2} R=\left[2^{2}+\left(\frac{2.65}{\sqrt{2}}\right)^{2}+\left(\frac{2.43}{\sqrt{2}}\right)^{2}\right] 5=52.2 \mathrm{W}
$$

__Example 10 Sinusoidal Source and a Nonlinear Load__

A sinusoidal voltage source of $v(t)=100 \cos (377 t) \mathrm{V}$ is applied to a nonlinear load, resulting in a nonsinusoidal current which is expressed in Fourier series form as

$$
i(t)=8+15 \cos \left(377 t+30^{\circ}\right)+6 \cos \left[2(377) t+45^{\circ}\right]+2 \cos \left[3(377) t+60^{\circ}\right]
$$

Determine __(a)__ the power absorbed by the load, __(b)__ the power factor of the load, __(c)__ the distortion factor of the load current, __(d)__ the total harmonic distortion of the load
current.

> __■ Solution__
> 
> __(a)__ The power absorbed by the load is determined by computing the power absorbed at each frequency in the Fourier series [Eq. 59].
> 
> $$
\begin{array}{l}
P=(0)(8)+\left(\frac{100}{\sqrt{2}}\right)\left(\frac{15}{\sqrt{2}}\right) \cos 30^{\circ}+(0)\left(\frac{6}{\sqrt{2}}\right) \cos 45^{\circ}+(0)\left(\frac{2}{\sqrt{2}}\right) \cos 60^{\circ} \\
P=\left(\frac{100}{\sqrt{2}}\right)\left(\frac{15}{\sqrt{2}}\right) \cos 30^{\circ}=650 \mathrm{W}
\end{array}
$$
>
> __(b)__ The rms voltage is
> $$
V_{\mathrm{ms}}=\frac{100}{\sqrt{2}}=70.7 \mathrm{V}
$$
> 
> and the rms current is computed from [Eq. 64]:
> 
> $$
I_{\mathrm{Fms}}=\sqrt{8^{2}+\left(\frac{15}{\sqrt{2}}\right)^{2}+\left(\frac{6}{\sqrt{2}}\right)^{2}+\left(\frac{2}{\sqrt{2}}\right)^{2}}=14.0 \mathrm{A}
$$
> 
> The power factor is
> 
> $$
\mathrm{pf}=\frac{P}{S}=\frac{P}{V_{\mathrm{rms}} I_{\mathrm{rms}}}=\frac{650}{(70.7)(14.0)}=0.66
$$
> 
> Alternatively, power factor can be computed from [Eq. 63]:
> 
> $$
\mathrm{pf}=\frac{I_{1 . \mathrm{ms}} \cos \left(\theta_{1}-\phi_{1}\right)}{I_{\mathrm{ms}}}=\frac{\left(\frac{15}{\sqrt{2}}\right) \cos \left(0-30^{\circ}\right)}{14.0}=0.66
$$
> 
> __(c)__ The distortion factor is computed from [Eq. 65] as
> 
> $$
\mathrm{DF}=\frac{I_{1, \mathrm{mms}}}{I_{\mathrm{ms}}}=\frac{\frac{15}{\sqrt{2}}}{14.0}=0.76
$$
> 
> __(d)__ The total harmonic distortion of the load current is obtained from [Eq. 68].
> 
> $$
\mathrm{THD}=\sqrt{\frac{I_{\mathrm{mls}}^{2}-I_{1, \mathrm{rms}}^{2}}{I_{1, \mathrm{ms}}^{2}}}=\sqrt{\frac{14^{2}-\left(\frac{15}{\sqrt{2}}\right)^{2}}{\left(\frac{15}{\sqrt{2}}\right)^{2}}}=0.86=86 \%
$$
