# Flyback Converter Portfolios 

__Example 07001__ Flyback Converter
A flyback converter of Fig. 07002 has the following circuit parameters:

$$
\begin{array}{l}
V_{s}=24 \mathrm{V} \\
N_{1} / N_{2}=3.0 \\
L_{m}=500 \mu \mathrm{H} \\
R=5 \Omega \\
C=200 \mu \mathrm{F} \\
f=40 \mathrm{kHz} \\
V_{o}=5 \mathrm{V}
\end{array}
$$

Determine ( $a$ ) the required duty ratio $D ;(b)$ the average, maximum, and minimum values for the current in $L_{m} ;$ and $(c)$ the output voltage ripple. Assume that all components are ideal.

   * __Solution__
>
> (a) Rearranging [Eq. 07004] yields
> 
> $$
\begin{array}{l}
V_{o}=V_{s}\left(\frac{D}{1-D}\right)\left(\frac{N_{2}}{N_{1}}\right) \\
D=\frac{1}{\left(V_{s} / V_{o}\right)\left(N_{2} / N_{1}\right)+1}=\frac{1}{(24 / 5)(1 / 3)+1}=0.385
\end{array}
$$
> 
> (b) Average current in $L_{m}$ is determined from [Eq. 07008].
> 
> $$
I_{L_{m}}=\frac{V_{o}^{2}}{V_{s} D R}=\frac{5^{2}}{(24)(0.385)(5)}=540 \mathrm{mA}
$$
> 
> The change in $i_{L_{m}}$ can be calculated from [Eq. 07002].
> 
> $$
\Delta i_{L_{m}}=\frac{V_{s} D}{L_{m} f}=\frac{(24)(0.385)}{500(10)^{-6}(40,000)}=460 \mathrm{mA}
$$
> 
> Maximum and minimum inductor currents can be computed from
> 
> $$
\begin{array}{l}
I_{L_{m, \max }}=I_{L_{m}}+\frac{\Delta i_{L_{m}}}{2}=540+\frac{460}{2}=770 \mathrm{mA} \\
I_{L_{m, \operatorname{mith}}}=I_{L_{m}}-\frac{\Delta i_{L_{n}}}{2}=540-\frac{460}{2}=310 \mathrm{mA}
\end{array}
$$
> 
> The [Eq. 07010] and [Eq. 07011], which are derived from the above computation, could also be used directly to obtain the maximum and minimum currents. Note that a positive $I_{L m, \min }$ verifies continuous current in $L_{m}$.
> 
> (c) Output voltage ripple is computed from [Eq. 07014].
> 
> $$
\frac{\Delta V_{o}}{V_{n}}=\frac{D}{R C f}=\frac{0.385}{(5)\left[200(10)^{-6}\right](40,000)}=0.0096=0.96 \%
$$

__Example 07002__ Flyback Converter Design, Continuous-Current Mode
Design a converter to produce an output voltage of $36 \mathrm{V}$ from a $3.3 \mathrm{~V}$ source. The output current is $0.1 \mathrm{~A}$. Design for an output ripple voltage of 2 percent. Include ESR when choosing a capacitor. Assume for this problem that the ESR is related to the capacitor value by $r_{C}=10^{-5} / C$

   * __Solution__
>
> Considering a boost converter for this application and calculating the required duty ratio from [Eq. 06027]
> 
> $$
D=1-\frac{V_{s}}{V_{o}}=1-\frac{3.3}{36}=0.908
$$
> 
> The result of a high duty ratio will likely be that the converter will not function as desired because of losses in the circuit (Fig. 06010). Therefore, a boost converter would not be a good choice. A flyback converter is much better suited for this application.
> 
> As a somewhat arbitrary design decision, start by letting the duty ratio be $0.4$. From [Eq. 07004], the transformer turns ratio is calculated to be
> 
> $$
\left(\frac{N_{2}}{N_{1}}\right)=\frac{V_{o}}{V_{s}}\left(\frac{1-D}{D}\right)=\frac{36}{3.3}\left(\frac{1-0.4}{0.4}\right)=16.36
$$
> 
> Rounding, let $N_{2} / N_{1}=16$. Recalculating the duty ratio using a turns ratio of 16 gives $D = 0.405$.
> 
> To determine $L_{m}$, first compute the average current in $L_{m}$ from [Eq. 07009], using $I_{o}=V_{o} / R$
> 
> $$
I_{L_{m}}=\frac{V_{o}}{(1-D) R}\left(\frac{N_{2}}{N_{1}}\right)=\frac{I_{o}}{1-D}\left(\frac{N_{2}}{N_{1}}\right)=\left(\frac{0.1}{1-0.405}\right) 16=2.69 \mathrm{A}
$$
> 
> Let the current variation in $L_{m}$ be 40 percent of the average current: $\Delta i_{l_{n}}=0.4(2.69)=1.08 \mathrm{~A}$. As another somewhat arbitrary choice, let the switching frequency be $100 \mathrm{KHz}$. Using [Eq. 070013]
> 
> $$
L_{m}=\frac{V_{s} D}{\Delta i_{L_{m}} f}=\frac{3.3(0.405)}{1.08(100,000)}=12.4 \mu \mathrm{H}
$$
> 
> Maximum and minimum currents in $L_{m}$ are found from [Eqs. 07010] and [07011] as $3.23$ and $2.15 \mathrm{~A}$, respectively.
> 
> The output voltage ripple is to be limited to 2 percent, which is $0.02(36)=0.72 \mathrm{V}$ Assume that the primary cause of the voltage ripple will be the voltage drop across the equivalent series resistance $\Delta i_{C^{T}}$. The peak-to-peak variation in capacitor current is the same as in the diode and the transformer secondary and is related to current in $L_{m}$ by
> 
> $$
\Delta i_{C}=I_{L_{\pi, \max }}\left(\frac{N_{1}}{N_{2}}\right)=(3.23 \mathrm{A})\left(\frac{1}{16}\right)=0.202 \mathrm{A}
$$
> 
> Using [Eq. 07015]
> 
> $$
r_{C}=\frac{\Delta V_{o, \mathrm{ESR}}}{\Delta i_{C}}=\frac{0.72 \mathrm{V}}{0.202 \mathrm{A}}=3.56 \Omega
$$
> 
> Using the relationship between ESR and capacitance given in this problem,
> $$
C=\frac{10^{-5}}{r_{C}}=\frac{10^{-5}}{3.56}=2.8 \mu \mathrm{F}
$$
> 
> The ripple voltage due to the capacitance only is obtained from [Eq. 07014] as
> 
> $$
\frac{\Delta V_{o}}{V_{o}}=\frac{D}{R C f}=\frac{0.405}{(36 \mathrm{V} / 0.1 \mathrm{A})\left[2.8(10)^{-6}\right](100,000)}=0.004=0.04 \%
$$
> 
> showing that the assumption that the voltage ripple is primarily due to the ESR was correct. A standard value of $3.3 \mu \mathrm{F}$ would be a good choice. Note that the designer should consult manufacturers' specifications for ESR when selecting a capacitor.
> 
> The turns ratio of the transformer, current variation, and switching frequency were selected somewhat arbitrarily, and many other combinations are suitable.
