### EXAMPLE 08001 Square-Wave Inverter with RL Load

The full-bridge inverter of Fig. 08001 has a switching sequence that produces a square wave voltage across a series RL load. The switching frequency is $60 \mathrm{Hz}$, $V_{\mathrm{dc}}=100 \mathrm{V}$, $R=10 \Omega$ 	, and$L=25 \mathrm{mH}$. Determine (a) an expression for load current, (b) the power absorbed by the load, and (c) the average current in the dc source.

> Solution
>  
> (a) From the parameters given,
>  
> $T=1 / f=1 / 60=0.0167 \mathrm{s}$
>  
> $\tau=L / R=0.025 / 10=0.0025 \mathrm{s}$
>  
> $T / 2 \tau=3.33$
>  
> The [Eq. 08008] is used to determine the maximum and minimum current.
>  
> $$
I_{\max }=-I_{\min }=\frac{100}{10}\left(\frac{1-e^{-3.33}}{1+e^{-3.33}}\right)=9.31 \mathrm{A}
$$
>  
> [Equation 5] is then evaluated to give load current.
>  
> $$
\begin{aligned} i_{o}(t) &=\frac{100}{10}+\left(-9.31-\frac{100}{10}\right) e^{-t / 0.0025} \\ &=10-19.31 e^{-t / 0.0025} \quad 0 \leq t \leq \frac{1}{120} \end{aligned}
$$
>  
> $$
\begin{aligned} i_{o}(t) &=-\frac{100}{10}+\left(9.31+\frac{100}{10}\right) e^{-(t-0.0167 / 2) / 0.0025} \\ &=-10+19.31 e^{-(t-0.00835) 0.0025} \quad \frac{1}{120} \leq t \leq \frac{1}{60} \end{aligned}
$$
>  
> (b) Power is computed from $I_{\mathrm{ms}}^{2} R$, where $I_{\mathrm{ms}}$ is computed from [Eq. 08009].
>  
> $$
I_{\mathrm{ms}}=\sqrt{\frac{1}{120} \int_{0}^{1 / 120}\left[(10-19.31) e^{-t / 0.0025}\right]^{2} d t}=6.64 \mathrm{A}
$$
>  
> Power absorbed by the load is
>  
>  $$
P=I_{\mathrm{rms}}^{2} R=(6.64)^{2}(10)=441 \mathrm{W}
$$
>  
> (c) Average source current can also be computed by equating source and load power, assuming a lossless converter. Using [Eq. 08010],
>  
> $$
I_{s}=\frac{P_{\mathrm{dc}}}{V_{\mathrm{dc}}}=\frac{441}{100}=4.41 \mathrm{A}
$$
>  
> Average power could also be computed from the average of the current
expression in part (a).

__EXAMPLE 08002__ Fourier Series Analysis
For the inverter in Example 08001 ($V_{\mathrm{dc}}=100 \mathrm{V}, R=10 \Omega, L=25 \mathrm{mH}, f=60 \mathrm{Hz}$), determine the amplitudes of the Fourier series terms for the square wave load voltage, the amplitudes of the Fourier series terms for load current, and the power absorbed by the load.

> Solution
> The load voltage is represented as the Fourier series in [Eq. 080016]. The amplitude of each voltage term is
> 
> $$
V_{n}=\frac{4 V_{d c}}{n \pi}=\frac{4(400)}{n \pi}
$$
> 
> The amplitude of each current term is determined from [Eq. 08014],
> 
> $$
I_{n}=\frac{V_{n}}{Z_{n}}=\frac{V_{n}}{\sqrt{R^{2}+\left(n \omega_{0} L\right)^{2}}}=\frac{4(400) / n \pi}{\sqrt{10^{2}+[n(2 \pi 60)(0.025)]^{2}}}
$$
> 
> Power at each frequency is determined from [Eq. 08015].
> 
> $$
P_{n}=I_{n, \mathrm{ms}}^{2} R=\left(\frac{I_{n}}{\sqrt{2}}\right)^{2} R
$$
> 
>

> $$
\begin{array}{lrrrrr}
\hline n & f_{n}(\mathrm{Hz}) & V_{n}(\mathrm{V}) & Z_{n}(\Omega) & I_{n}(\mathrm{A}) & P_{n}(\mathrm{W}) \\
\hline 1 & 60 & 127.3 & 13.7 & 9.27 & 429.3 \\
3 & 180 & 42.4 & 30.0 & 1.42 & 10.0 \\

> 
> Table 08001 summarizes the Fourier series quantities for the circuit of Example 08001. As the harmonic number n increases, the amplitude of the Fourier voltage component decreases and the magnitude of the corresponding impedance increases, both resulting in small currents for higher-order harmonics. Therefore, only the first few terms of the series are of practical interest. Note how the current and power terms become vanishingly small for all but the first few frequencies. 
> Power absorbed by the load is computed from [Eq. 08015].
> 
> $$
P=\sum P_{n}=429.3+10.0+1.40+0.37+0.14+\cdots \approx 441 \mathrm{W}
$$
> 
> which agrees with the result in Example 08001.

__EXAMPLE 08003__ The total harmonic distortion
Determine the total harmonic distortion of the load voltage and the load current for the square-wave inverter in Examples 08001 and 08002.

> Solution
> Use the Fourier series for the square wave in [Eq. 08016] and the definition of THD in [Eq. 08017]. The rms value of the square wave voltage is the same as the peak value, and the fundamental frequency component is the first term in [Eq. 08016],
> 
> $$
V_{\mathrm{rms}}=V_{\mathrm{dc}}
$$
> <br><br/>
> 
> $$
V_{1, \mathrm{ms}}=\frac{V_{1}}{\sqrt{2}}=\frac{4 V_{\mathrm{dc}}}{\sqrt{2} \pi}
$$
> 
> Using [Eq. 08017] to compute the total harmonic distortion for voltage,
> 
> $$
\mathrm{THD}_{V}=\frac{\sqrt{V_{\mathrm{ms}}^{2}-V_{1, \mathrm{rms}}^{2}}}{V_{\mathrm{I}, \mathrm{ms}}}=\frac{\sqrt{V_{\mathrm{dc}}^{2}-\left(4 V_{\mathrm{dc}} / \sqrt{2} \pi\right)^{2}}}{4 V_{\mathrm{dc}} / \sqrt{2} \pi}=0.483=48.3 \%
$$
> 
> The THD of the current is computed using the truncated Fourier series which was determined in Example 08002.
> 
> <br><br/>
> $$
\begin{aligned} \mathrm{THD}_{I} &=\frac{\sqrt{\sum_{n=2}^{\infty}\left(I_{n, \operatorname{mis}}\right)^{2}}}{I_{1, \mathrm{ms}}} \\ & \approx \frac{\sqrt{(1.42 / \sqrt{2})^{2}+(0.53 / \sqrt{2})^{2}+(0.27 / \sqrt{2})^{2}+(0.17 / \sqrt{2})^{2}}}{9.27 / \sqrt{2}}=0.167=16.7 \% \end{aligned}
$$

__Example 08006__ A Two-Source Multilevel Inverter
For the two-source multilevel inverter of Fig. 08009 with $V_{\mathrm{dc}}=100 \mathrm{V}:(a)$ Determine the Fourier coefficients through $n=9$ and the modulation index for $\alpha_{1}=20^{\circ}$ and $\alpha_{2}=40^{\circ}$
(b) Determine $\alpha_{1}$ and $\alpha_{2}$ such that the third harmonic $(n=3)$ is eliminated and $M_{t}=0.8$

> Solution
Using [Eq. 08024] to evaluate the Fourier coefficients,
> 
> $$
V_{n}=\frac{4 V_{\mathrm{dc}}}{n \pi}\left[\cos \left(n \alpha_{1}\right)+\cos \left(n \alpha_{2}\right)\right]=\frac{4(100)}{n \pi}\left[\cos \left(n 20^{\circ}\right)+\cos \left(n 40^{\circ}\right)\right]
$$
> 
> resulting in $V_{1}=217$, $V_{3}=0$, $V_{5}=-28.4$, $V_{7}=-10.8$, and $V_{9}=0$. Note that the third and ninth harmonics are eliminated. The even harmonics are not present. The modulation index $M_{i}$ is evaluated from [Eq. 08025].
> 
> $$
M_{i}=\frac{\cos \alpha_{1}+\cos \alpha_{2}}{2}=\frac{\cos 20^{\circ}+\cos 40^{\circ}}{2}=0.853
$$
> 
> The amplitude of the fundamental frequency voltage is therefore 85.3 percent of that of a square wave of $\pm 200 \mathrm{V}$.
(b) To achieve simultaneous elimination of the third harmonic and a modulation index of $M_{t}=0.8$ requires the solution to the equations
> 
> $$
\cos \left(3 \alpha_{1}\right)+\cos \left(3 \alpha_{2}\right)=0
$$
> 
> and
> $$
\cos \left(\alpha_{1}\right)+\cos \left(\alpha_{2}\right)=2 M_{i}=1.6
$$
> 
> Using an iterative method, $\alpha_{1}=7.6^{\circ}$ and $\alpha_{2}=52.4^{\circ}$
> 

__Exaple 08007__ A Five-Source Multilevel Inverter
Determine the delay angles required for a five-source cascade multilevel converter that will eliminate harmonics $5,7,11,$ and 13 and will have a modulation index $M_{t}=0.8$.

> Solution 
> The delay angles must satisfy these simultaneous equations:
> 
> $$
\begin{array}{l}
\cos \left(5 \alpha_{1}\right)+\cos \left(5 \alpha_{2}\right)+\cos \left(5 \alpha_{3}\right)+\cos \left(5 \alpha_{4}\right)+\cos \left(5 \alpha_{5}\right)=0 \\
\cos \left(7 \alpha_{1}\right)+\cos \left(7 \alpha_{2}\right)+\cos \left(7 \alpha_{3}\right)+\cos \left(7 \alpha_{4}\right)+\cos \left(7 \alpha_{5}\right)=0 \\
\cos \left(11 \alpha_{1}\right)+\cos \left(11 \alpha_{2}\right)+\cos \left(11 \alpha_{3}\right)+\cos \left(11 \alpha_{4}\right)+\cos \left(11 \alpha_{5}\right)=0 \\
\cos \left(13 \alpha_{1}\right)+\cos \left(13 \alpha_{2}\right)+\cos \left(13 \alpha_{3}\right)+\cos \left(13 \alpha_{4}\right)+\cos \left(13 \alpha_{5}\right)=0 \\
\cos \left(\alpha_{1}\right)+\cos \left(\alpha_{2}\right)+\cos \left(\alpha_{3}\right)+\cos \left(\alpha_{4}\right)+\cos \left(\alpha_{5}\right)=5 M_{i}=5(0.8)=4
\end{array}
$$
> 
> An iteration method such as the Newton-Raphson method must be used to solve these equations. The result is $\alpha_{1}=6.57^{\circ}$, $ \alpha_{2}=18.94^{\circ}$, $\alpha_{3}=27.18^{\circ}$, $\alpha_{4}=45.14^{\circ}$, and $\alpha_{5}=62.24^{\circ}$. See the references in the Bibliography for information on the technique.

__Example 08008__ A PWM Inverter

The full-bridge inverter is used to produce a $60-\mathrm{Hz}$ voltage across a series $R L$ load using bipolar PWM. The dc input to the bridge is $100 \mathrm{V},$ the amplitude modulation ratio $m_{a}$ is $0.8,$ and the frequency modulation ratio $m_{f}$ is $21\left[f_{\mathrm{tri}}=(21)(60)=1260 \mathrm{Hz}\right]$. The load has a resistance of $R=10 \Omega$ and series inductance $L=20 \mathrm{mH}$. Determine $(a)$ the amplitude of the $60-\mathrm{Hz}$ component of the output voltage and load current, $(b)$ the power absorbed by the load resistor, and $(c)$ the $\mathrm{THD}$ of the load current.

> Solution
(a) Using [Eq. 08038] and Table 08003, the amplitude of the 60-Hz fundamental frequency is
> 
> $$
V_{1}=m_{a} V_{\mathrm{dc}}=(0.8)(100)=80 \mathrm{V}
$$
> 
> The current amplitudes are determined using phasor analysis:
> 
> $$
I_{n}=\frac{V_{n}}{Z_{n}}=\frac{V_{n}}{\sqrt{R^{2}+\left(n \omega_{0} L\right)^{2}}}
$$   [Eq. 08041]
> 
> For the fundamental frequency,
> 
> $$
I_{1}=\frac{80}{\sqrt{10^{2}+[(1)(2 \pi 60)(0.02)]^{2}}}=6.39 \mathrm{A}
$$
> 
> (b) With $m_{f}=21,$ the first harmonics are at $n=21,19,$ and $23$. Using Table 08003
> 
> $$
\begin{array}{l}
V_{21}=(0.82)(100)=82 \mathrm{V} \\
V_{19}=V_{23}=(0.22)(100)=22 \mathrm{V}
\end{array}
$$
> 
> Current at each of the harmonics is determined from [Eq. 08041] Power at each frequency is determined from
> 
> $$
P_{n}=\left(I_{n, \mathrm{rms}}\right)^{2} R=\left(\frac{I_{n}}{\sqrt{2}}\right)^{2} R
$$
>

> 
> $$

\hline 1 & 60 & 80.0 & 12.5 & 6.39 & 4.52 & 204.0 \\

>

$$
> 
> Higher-order harmonics contribute little power and can be neglected.
(c) The THD of the load current is determined using Eq. 08017 with the rms current of the harmonics approximated by the first few terms indicated in Table 08004
> 
> $$
\mathrm{THD}_{1}=\frac{\sqrt{\sum_{n=2}^{\infty} I_{n, \mathrm{rms}}^{2}}}{I_{1, \mathrm{mls}}} \approx \frac{\sqrt{(0.11)^{2}+(0.36)^{2}+(0.09)^{2}}}{4.52}=0.087=8.7 \%
$$

However, since the impedance of the load increases and the amplitudes of the harmonics generally decrease as $n$ increases, the above approximation should be acceptable. (Including currents through $n=100$ gives a THD of 9.1 percent.)

__Example 08009__ PWM Inverter Design

Design a bipolar PWM inverter that will produce a $75-V$ rms $60-\mathrm{Hz}$ output from a $150-V$ dc source. The load is a series $R L$ combination with $R=12 \Omega$ and $L=60 \mathrm{mH}$. Select the switching frequency such that the current THD is less than 10 percent.

> Solution 
The required amplitude modulation ratio is determined from Eq. 080038,
> 
> $$
m_{a}=\frac{V_{1}}{V_{d c}}=\frac{75 \sqrt{2}}{150}=0.707
$$
> 
> The current amplitude at $60 \mathrm{Hz}$ is
> 
> $$
I_{1}=\frac{V_{1}}{Z_{1}}=\frac{75 \sqrt{2}}{\sqrt{12^{2}+[(2 \pi 60)(0.06)]^{2}}}=4.14 \mathrm{A}
$$
> 
> The rms value of the harmonic current has a limit imposed by the required THD,
> 
> $$
\sqrt{\sum_{n=2}^{\infty}\left(I_{n, \operatorname{mn}}\right)^{2}}<0.1 I_{1, \mathrm{rms}}=0.1\left(\frac{4.14}{\sqrt{2}}\right)=0.293 \mathrm{A}
$$
> 
> The term that will produce the dominant harmonic current is at the switching frequency. As an approximation, assume that the harmonic content of the load current is the same as the dominant harmonic at the carrier frequency:
> 
> $$
\sqrt{\sum_{n=2}^{\infty}\left(I_{n, \mathrm{Fms}}\right)^{2}} \approx I_{m f, \mathrm{ms}}=\frac{I_{m f}}{\sqrt{2}}
$$
> 
> The amplitude of the current harmonic at the carrier frequency is then approximated as
> 
> $$
I_{m f}<(0.1)(4.14)=0.414 \mathrm{A}
$$
> 
> Table 08003 indicates that the normalized voltage harmonic for $n=m_{f}$ and for $m_{a}=0.7$ is $0.92 .$ The voltage amplitude for $n=m_{f}$ is then
> 
> $$
V_{m f}=0.92 V_{d c}=(0.92)(150)=138 \mathrm{V}
$$
> 
> The minimum load impedance at the carrier frequency is then
> 
> $$
Z_{m f}=\frac{V_{m f}}{I_{m f}}=\frac{138}{0.414}=333 \Omega
$$
> 
> Because the impedance at the carrier frequency must be much larger than the $12 \Omega$ load resistance, assume the impedance at the carrier frequency is entirely inductive reactance,
> 
> $$
Z_{m f} \approx \omega L=m_{f} \omega_{0} L
$$
> 
> For the load impedance to be greater than $333 \Omega$,
> 
> $$
\begin{array}{c}
m_{f} \omega_{0} L>333 \\
m_{f}>\frac{333}{(377)(0.06)}=14.7
\end{array}
$$
> 
> Selecting $m_{f}$ to be at least 15 would marginally meet the design specifications. However, the estimate of the harmonic content used in the calculations will be low, so a higher carrier frequency is a more prudent selection. Let $m_{f}=17$, which is the next odd integer. The carrier frequency is then
> 
> $$
f_{\mathrm{tr}}=m_{f} f_{\mathrm{ref}}=(17)(60)=1020 \mathrm{Hz}
$$   [Eq. 080]
> 
> Further increasing $m_{f}$ would reduce the current $T H D$, but at the expense of larger switching losses. A PSpice simulation, as discussed later in this chapter, can be used to verify that the design meets the specifications.
