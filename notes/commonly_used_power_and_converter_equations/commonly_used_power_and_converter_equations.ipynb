# Commonly used Power Equations



* Instantaneous power:
$$
p(t)=v(t) i(t)
$$   [Eq. 1.1]
<br><br/>

* Energy:
$$
W=\int_{t_{1}}^{t_{2}} p(t) d t
$$   [Eq. 1.2]
<br><br/>

* Average power:
$$
P=\frac{W}{T}=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} p(t) d t=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} v(t) i(t) d t
$$   [Eq. 1.3]
<br><br/>

* Average power for a dc voltage source:
$$
P_{\mathrm{dc}}=V_{\mathrm{dc}} I_{\mathrm{avg}}
$$   [Eq. 1.4]
<br><br/>

* rms voltage:
$$
V_{\mathrm{rms}}=\sqrt{\frac{1}{T} \int_{0}^{T} v^{2}(t) d t}
$$   [Eq. 1.5]
<br><br/>

* rms for a series of ortogonal components $v=v_{1}+v_{2}+v_{3}+\cdots$:
$$
V_{\mathrm{rms}}=\sqrt{V_{1, \mathrm{rms}}^{2}+V_{2, \mathrm{rms}}^{2}+V_{3, \mathrm{rms}}^{2}+\cdots}
$$   [Eq. 1.6]
<br><br/>

* rms current for a triangular wave:
$$
I_{rms }=\frac{I_{m}}{\sqrt{3}}
$$   [Eq. 1.7]
<br><br/>

* rms current for an offset triangular wave:
$$
I_{\mathrm{rms}}=\sqrt{\left(\frac{I_{\mathrm{m}}}{\sqrt{3}}\right)^{2}+I_{\mathrm{dc}}^{2}}
$$   [Eq. 1.8]
<br><br/>

* rms voltage for a sine wave or a full-wave rectified sine wave
$$
V_{rms }=\frac{V_{m}}{\sqrt{2}}
$$   [Eq. 1.9]
<br><br/>

* rms voltage for a half-wave rectified sine wave:
$$
V_{rms }=\frac{V_{m}}{2}
$$   [Eq. 1.10]
<br><br/>

* Power factor pf:
$$
\mathrm{pf}=\frac{P}{S}=\frac{P}{V_{\mathrm{rms}} I_{\mathrm{rms}}}
$$   [Eq. 1.11]
<br><br/>

* Total harmonic distortion THD:
$$
\mathrm{THD}=\frac{\sqrt{\sum_{n=2}^{\infty} I_{n}^{2}}}{I_{1}}
$$   [Eq. 1.12]
<br><br/>

* Distortion factor:
$$
\mathrm{DF}=\sqrt{\frac{1}{1+(\mathrm{THD})^{2}}}
$$   [Eq. 1.13]
<br><br/>

* Form factor:
$$
\text { Form factor }=\frac{I_{\mathrm{ms}}}{I_{\mathrm{avg}}}
$$   [Eq. 1.14]
<br><br/>

* Crest factor:
$$
\text { Crest factor }=\frac{I_{\text {peak }}}{I_{\text {rms }}}
$$   [Eq. 1.15]
<br><br/>

# Commonly used Converter Transfer Functions

* Buck converter:
$$
V_{o}=V_{s} D
$$   [Eq. 2.1]
<br><br/>

* Boost converter:
$$
V_{o}=\frac{V_{s}}{1-D}
$$   [Eq. 2.2]
<br><br/>


* Buck-boost and Â´Cuk converters:
$$
V_{o}=-V_{s}\left(\frac{D}{1-D}\right)
$$   [Eq. 2.3]
<br><br/>

* SEPIC:
$$
V_{o}=V_{s}\left(\frac{D}{1-D}\right)
$$   [Eq. 2.4]
<br><br/>

* Flyback converter:
$$
V_{o}=V_{s}\left(\frac{D}{1-D}\right)\left(\frac{N_{2}}{N_{1}}\right)
$$   [Eq. 2.5]
<br><br/>

* Forward converter: 
$$
V_{o}=V_{s} D\left(\frac{N_{2}}{N_{1}}\right)
$$   [Eq. 2.6]
<br><br/>
