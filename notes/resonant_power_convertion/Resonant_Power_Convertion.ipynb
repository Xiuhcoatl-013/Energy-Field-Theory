# Resonant Power Convertion





## INTRODUCTION
Imperfect switching is a major contributor to power loss in converters. Switching devices absorb power when they turn on or off if they go through a transition when both voltage and current are nonzero. As the switching frequency increases, these transitions occur more often and the average power loss in the device increases. High switching frequencies are otherwise
desirable because of the reduced size of filter components and transformers, which reduces the size and weight of the converter.

In resonant switching circuits, switching takes place when voltage and/or current is zero, thus avoiding simultaneous transitions of voltage and current and thereby eliminating switching losses. This type of switching is called soft
switching, as opposed to hard switching in circuits such as the buck converter.

Resonant converters include resonant switch converters, load resonant converters,and resonant dc link converters.

# The Series Resonant Inverter


The series resonant inverter (dc-to-ac converter) of Fig. 3a is one application of resonant converters. In a series resonant inverter, an inductor and a capacitor are placed in series with a load resistor. The switches produce a square wave voltage, and the inductor-capacitor combination is selected such that the resonant frequency is the same as the switching frequency.

The analysis begins by considering the frequency response of the $R L C$ circuit of Fig. 3b. The input and output voltage amplitudes are related by

$$\frac{V_{o}}{V_{i}}=\frac{R}{\sqrt{R^{2}+(\omega L-(1 / \omega C))^{2}}}=\frac{1}{\sqrt{1+((\omega L / R)-(1 / \omega R C))^{2}}}$$   [45]

Resonance is at the frequency

$$\omega_{0}=\frac{1}{\sqrt{L C}}$$   [46]

or

$$f_{0}=\frac{1}{2 \pi \sqrt{L C}}$$   [47]

At resonance, the impedances of the inductance and capacitance cancel, and the load appears as a resistance. If the bridge output is a square wave at frequency $f_{0},$ the $L C$ combination acts as a filter, passing the fundamental frequency and attenuating the harmonics. If the third and higher harmonics of the square wave bridge output are effectively removed, the voltage across the load resistor is essentially a sinusoid at the square wave's fundamental frequency.

The amplitude of the fundamental frequency of a square wave voltage of $\pm V_{\mathrm{dc}}$ is

$$r_{1}=\frac{4 V_{1}}{\pi}$$   [48]

The frequency response of the filter could be expressed in terms of bandwidth, which is also characterized by the quality factor $Q .$

$$Q=\frac{\omega_{0} L}{R}=\frac{1}{\omega_{0} R C}$$   [49]

[Eq. 45] can be expressed in terms of $\omega_{0}$ and $Q$ :

$$\frac{V_{o}}{V_{i}}=\frac{1}{\sqrt{1+Q^{2}\left(\left(\omega / \omega_{0}\right)-\left(\omega_{0} / \omega\right)\right)^{2}}}$$   [50]

The normalized frequency response with $Q$ as a parameter is shown in Fig. 3c. The total harmonic distortion THD, of the voltage across the load resistor is reduced by increasing the $Q$ of the filter. Increasing inductance and reducing capacitance increase $Q$


## Switching Losses
An important feature of the resonant inverter is that switch losses are reduced over that of the hard swithcing inverters discussed earlier. If switching is at the resonant frequency and the $Q$ of the circuit is high, the switches operate when the load current is at or near zero. This is significant because the power absorbed by the switches is less than in the nonresonant inverter.

## Amplitude Control
If the frequency of the load voltage is not critical, the amplitude of the fundamental frequency across the load resistor can be controlled by shifting the switching frequency off of resonance. Power absorbed by the load resistor is thus controlled by the switching frequency. Induction heating is an application.

The switching frequency should be shifted higher than resonance rather than lower when controlling the output. Higher switching frequencies moves the harmonics of the square wave higher, increasing the filter's effectiveness in removing them. Conversely, shifting the frequency lower than resonance moves the harmonics, particularly the third harmonic, closer to resonance and increases their amplitudes in the output.

# [EXAMPLE] A Resonant Inverter 

A $10 \Omega$ resistive load requires a $1000 \mathrm{Hz}, 50 \mathrm{V}rms$ sinusoidal voltage. The THD of the load voltage must be no more than 5 percent. An adjustable dc source is available. (a) Design an inverter for this application. (b) Determine the maximum voltage across the capacitor. (c) Verify the design with a PSpice simulation.

### Solution

(a) The full-bridge converter of Fig. 3a with $1000 \mathrm{Hz}$ square wave switching and series resonant $L C$ filter is selected for this design. The amplitude of a $50 \mathrm{V}rms$ sinusoidal voltage is $\sqrt{2}(50)=70.7 \mathrm{V}$. The required de input voltage is determined from [Eq. 48].

$$\begin{aligned}
70.7 &=\frac{4 V_{\mathrm{dc}}}{\pi} \\
V_{\mathrm{dc}} &=55.5 \mathrm{V}
\end{aligned}$$


The resonant frequency of the filter must be $1000 \mathrm{Hz}$, establishing the $L C$ product. The $Q$ of the filter and the THD limit are used to detemine the values of $L$ and $C$. The third harmonic of the square wave is the largest and will be the least attenuated by the filter. Estimating the THD from the third harmonic,

$$\mathrm{THD}=\frac{\sqrt{\sum_{n \neq 1} V_{n}^{2}}}{V_{1}} \approx \frac{V_{3}}{V_{1}}$$

where $V_{1}$ and $V_{3}$ are the amplitudes of the fundamental and third harmonic frequencies, respectively, across the load. Using the foregoing approximation, the amplitude of the third harmonic of the load voltage must be at most

$$V_{3}<(\mathrm{THD})\left(V_{1}\right)=(0.05)(70.7)=3.54 \mathrm{V}$$


For the square wave, $V_{3}=V_{1} / 3=70.7 / 3 .$ Using [Eq. 50], $Q$ is determined from the magnitude of the third harmonic output with the third harmonic input, $70.7 / 3,$ at $\omega=3 \omega_{0}$

$$ \frac{V_{o, 3}}{V_{l, 3}}=\frac{3.54}{70.7 / 3}=\sqrt{\frac{1}{1+Q^{2}\left(\left(3 \omega_{0} / \omega_{0}\right)-\left(\omega_{0} / 3 \omega_{0}\right)\right)^{2}}}$$

Solving the preceding equation for $Q$ results in $Q=2.47$. Using [Eq. 49]

$$\begin{array}{l}
L=\frac{Q R}{\omega_{0}}=\frac{(2.47)(10)}{2 \pi(1000)}=3.93 \mathrm{mH} \\
C=\frac{1}{Q \omega_{0} R}=\frac{1}{(2.47)(2 \pi)(1000)(10)}=6.44 \mu \mathrm{F}
\end{array}$$

Power delivered to the load resistor at the fundamental frequency is $V_{\min }^{2} / R=50^{2} / 10 = 250 W$. Power delivered to the load at the third hamonic is $\left(2.5^{2}\right) / 10=0.63 \mathrm{W}$ showing that power at the harmonic frequencies is negligible.



(b) Voltage across the capacitor is estimated from phasor analysis at the fundamental frequency:

$$V_{C}=\left|\frac{I}{j \omega_{0} C}\right|=\frac{V_{1} / R}{\omega_{0} C}=\frac{70.7 / 10}{(2 \pi)(1000)(6.44)\left(10^{-6}\right)}=175 \mathrm{V}$$

At resonance, the inductor has the same impedance magnitude as the capacitor, so its voltage is also $175 \mathrm{V}$. The inductor and capacitor voltages would be larger if $Q$ were increased. Note that these voltages are larger than the output or source voltage.



(c) One method of doing a PSpice simulation is to use a square wave voltage as the input to the $R L C$ circuit. This assumes that the switching is ideal, but it is a good starting point to verify that the design meets the specifications. The circuit is shown in Fig. 4a. Output begins after three periods (3 ms) to allow steady-state conditions to be reached. The Probe output showing input and output voltages is seen in Fig. 4b and a Fourier analysis (FFT) from Probe is shown in Fig. 4c.

# A RESONANT SWITCH CONVERTER:
# ZERO-VOLTAGE SWITCHING


## Basic Operation

The circuit of Fig. 2a shows a method for using the oscillations of an $L C$ circuit for switching at zero voltage. The analysis assumes that the output filter produces a ripple-free current $I_{o}$ in $L_{o}$. Beginning with the switch closed, the current in the switch and in $L_{r}$ is $I_{o},$ the currents in $D_{1}$ and $D_{s}$ are zero, and the voltage across $C_{r}$ and the switch is zero.

The switch is opened (with zero voltage across it), and $i_{L}=I_{o}$ flows through the capacitor $C_{r},$ causing $v_{C}$ to increase linearly (Fig. 2b). When $v_{C}$ reaches the source voltage $V_{s},$ the diode $D_{1}$ becomes forward-biased, in effect forming a series circuit with $V_{s}, C_{r},$ and $L_{r}$ as shown in Fig. 2c. At this time, $i_{L}$ and $v_{C}$ in this underdamped series circuit begin to oscillate.

When $v_{C}$ returns to zero, diode $D_{s}$ turns on to carry $i_{l},$ which is negative (Fig. 2d). The voltage across $L_{r}$ is $V_{s}$, causing $i_{L}$ to increase linearly. The switch should be closed just after $D_{s}$ turns on for zero-voltage turn-on. When $i_{L}$ becomes positive, $D_{s}$ turns off and $i_{L}$ is carried by the switch. When $i_{L}$ reaches
$I_{o}, D_{1}$ turns off, and circuit conditions are back at the starting point. The analysis for each circuit condition is given next.




**Analysis for $0 \leq t \leq t_{1}$** The switch is opened at $t=0$. The capacitor current is then $I_{o}$ (Fig. 2b), causing the capacitor voltage, initially zero, to increase linearly. The voltage across $C_{r}$ is

$$v_{C}(t)=\frac{1}{C_{r}} \int_{0}^{t} I_{o} d \lambda=\frac{I_{o}}{C_{r}} t$$   [22]

The voltage across $L_{r}$ is zero because inductor current is $I_{o},$ which is assumed to be constant. The voltage at the filter input $v_{x}$ is

$$v_{x}(t)=V_{s}-v_{C}(t)=V_{s}-\frac{I_{o}}{C_{r}} t$$   [23]

which is a linearly decreasing function beginning at $V_{s}$. At $t=t_{1}, v_{x}=0$ and the diode turns on. Solving the preceding equation for $t_{1}$

$$t_{1}=\frac{V_{s} C_{r}}{I_{o}}$$   [24]

Eq. 23 can then be expressed as

$$v_{x}(t)=V_{s}\left(1-\frac{t}{t_{1}}\right)$$   [25]



**Analysis for $t_{1} \leq t \leq t_{2} \quad$** Diode $D_{1}$ is forward-biased and has $0 \mathrm{V}$ across it, and the equivalent circuit is shown in Fig. 2c. Kirchhoff's voltage law is expressed as

$$L_{r} \frac{d i_{L}(t)}{d t}+v_{C}(t)=V_{s}$$   [26]

Differentiating,

$$L_{r} \frac{d^{2} i_{L}(t)}{d t^{2}}+\frac{d v_{C}(t)}{d t}=0$$   [27]

Capacitor current is related to voltage by

$$\frac{d v_{C}(t)}{d t}=\frac{i_{C}(t)}{C_{r}}$$   [28]


Since inductor and capacitor currents are the same in this time interval, Eq. 27 can be expressed as

$$\frac{d^{2} i_{L}(t)}{d t^{2}}+\frac{i_{L}(t)}{L_{r} C_{r}}=0$$   [29]

Solving the preceding equation for $i_{L}$ by using the initial condition $i_{L}\left(t_{1}\right)=I_{o}$

$$i_{L}(t)=I_{o} \cos \left[\omega_{0}\left(t-t_{1}\right)\right]$$   [30]

where

$$\omega_{0}=\frac{1}{\sqrt{L_{r} C_{r}}}$$   [31]

Capacitor voltage is expressed as

$$v_{C}(t)=\frac{1}{C_{r}} \int_{t_{1}}^{t} i_{C}(\lambda) d \lambda+v_{C}\left(t_{1}\right)=\frac{1}{C_{r}} \int_{t_{1}}^{t} I_{o} \cos \left[\omega_{0}\left(\lambda-t_{1}\right)\right] d \lambda+V_{s}$$

which simplifies to

$$v_{C}(t)=V_{s}+I_{o} Z_{0} \sin \left[\omega_{0}\left(t-t_{1}\right)\right]$$   [32]

where

$$Z_{0}=\sqrt{\frac{L_{r}}{C_{r}}}$$   [33]

Note that the peak capacitor voltage is

$$V_{C, \text { peak }}=V_{s}+I_{o} Z_{0}=V_{s}+I_{o} \sqrt{\frac{L_{r}}{C_{r}}}$$   [34]

which is also the maximum reverse voltage across diode $D_{s}$ and is larger than the source voltage. 
With diode $D_{1}$ forward-biased, 

$$v_{x}=0$$   [35]. 

The diode $D_{s}$ across $C_{r}$ prevents $v_{C}$ from going negative, so Eq. 32 is valid for $v_{C}>0$. Solving Eq. 32 for the time $t=t_{2}$ when $v_{C}$ returns to zero,

$$t_{2}=\frac{1}{\omega_{0}}\left[\sin ^{-1}\left(\frac{-V_{s}}{I_{o} Z_{0}}\right)\right]+t_{1}$$

which can be expressed as

$$t_{2}=\frac{1}{\omega_{0}}\left[\sin ^{-1}\left(\frac{V_{s}}{I_{o} Z_{0}}\right)+\pi\right]+t_{1}$$   [36]

At $t=t_{2},$ diode $D_{s}$ turns on.



**Analysis for $t_{2} \leq t \leq t_{3}$** (Fig. 2d) After $t_{2},$ both diodes are forward-biased (Fig. 2d), the voltage across $L_{r}$ is $V_{s}$, and $i_{L}$, increases linearly until it reaches $I_{o}$ at $t_{3}$. The switch is reclosed just after $t_{2}$ when $v_{C}=0$ (zero-voltage turn-on) and the diode is on to carry a negative $i_{L}$. The current $i_{L}$ in the interval from $t_{2}$ to $t_{3}$ is expressed as

$$i_{L}(t)=\frac{1}{L_{r}} \int_{t_{2}}^{t} V_{s} d \lambda+i_{L}\left(t_{2}\right)=\frac{V_{s}}{L_{r}}\left(t-t_{2}\right)+I_{o} \cos \left[\omega_{0}\left(t_{2}-t_{1}\right)\right]$$   [37]

where $i_{L}\left(t_{2}\right)$ is from Eq. 30. Current at $t_{3}$ is $I_{o}$:

$$i_{L}\left(t_{3}\right)=I_{o}=\frac{V_{s}}{L_{r}}\left(t_{3}-t_{2}\right)+I_{o} \cos \left[\omega_{0}\left(t_{2}-t_{1}\right)\right]$$   [38]

Solving for $t_{3}$

$$t_{3}=\left(\frac{L_{r} I_{o}}{V_{s}}\right)\left\{1-\cos \left[\omega_{0}\left(t_{2}-t_{1}\right)\right]\right\}+t_{2}$$   [39]

Voltage $v_{x}$ is zero in this interval:

$$v_{x}=0$$   [40]

At $t=t_{3},$ diode $D_{1}$ turns on.



**Analysis for $t_{3} \leq t \leq T \quad$** In this interval, the switch is closed, both diodes are off, the current in the switch is $I_{o}$, and

$$v_{x}=V_{s}$$   [41]

The circuit remains in this condition until the switch is reopened. The time interval $T-t_{3}$ is determined by the switching frequency of the circuit. All other time intervals are determined by other circuit parameters.


**Output Voltage**

The voltage $v_{x}(t)$ at the input of the output filter is shown in Fig. 2f. Summarizing Eqs. 25, 35, 40 and 41

$$v_{x}(t)=\left\{\begin{array}{ll}
V_{s}\left(1-\frac{t}{t_{1}}\right) & 0<t<t_{1} \\
0 & t_{1}<t<t_{3} \\
V_{s} & t_{3}<t<T
\end{array}\right.$$   [42]

The output voltage is the average of $v_{x}(t) .$ Output voltage is

$$\begin{aligned}
V_{o} &=\frac{1}{T} \int_{0}^{T} v_{x} d t=\frac{1}{T}\left[\int_{0}^{t_{1}} V_{s}\left(1-\frac{t}{t_{1}}\right) d t+\int_{t_{3}}^{T} V_{s} d t\right] \\
&=\frac{V_{s}}{T}\left[\frac{t_{1}}{2}+\left(T-t_{3}\right)\right]
\end{aligned}$$   [43]

using $f_{s}=1 / T$

$$V_{o}=V_{s}\left[1-f_{s}\left(t_{3}-\frac{t_{1}}{2}\right)\right]$$   [44]

Times $t_{1},$ and $t_{3}$ in the preceding equation are determined from the circuit parameters as described by Eqs. 24,36 and 39. The output voltage is controlled by changing the switching frequency. The time interval when the switch is open is fixed, and the time interval when the switch is closed is varied. Times $t_{1}$ and $t_{3}$ are determined in part by the load current $I_{o}$, so output voltage is a function of load. Increasing the switching frequency decreases the time interval $T-t_{3}$ and thus reduces the output voltage. Normalized output voltage vs. switching frequency with the parameter $r=R_{L} / Z_{0}$ is shown in the graph in Fig. 2g. Output voltage is less than input voltage, as was the case for the buck converter.







### EXAMPLE

Resonant Switch Converter: Zero-Voltage Switching In the circuit of Fig. 2a

$$\begin{aligned}
&V_{s}=20 \mathrm{V}\\
&L_{r}=1 \mu \mathrm{H}\\
&C_{r}=0.047 \mu \mathrm{F}\\
&I_{o}=5 \mathrm{A}
\end{aligned}$$

(a) Determine the switching frequency such that the output voltage is $10 \mathrm{V}$.
(b) Determine the peak voltage across $D_{s}$ when it is reverse-biased.



### Solution
(a) From the circuit parameters,

$$\begin{aligned}
\omega_{0} &=\frac{1}{\sqrt{(10)^{-6}(0.047)(10)^{-6}}}=4.61\left(10^{6}\right) \mathrm{rad} / \mathrm{s} \\
Z_{0} &=\sqrt{\frac{L_{r}}{C_{r}}}=\sqrt{\frac{10^{-6}}{0.047\left(10^{-6}\right)}}=4.61 \Omega
\end{aligned}$$


Using Eq. 24 to solve for $t_{1}$

$$t_{1}=\frac{V_{s} C_{r}}{I_{o}}=\frac{(20)(0.047)\left(10^{-6}\right)}{5}=0.188 \mu \mathrm{s}$$

From Eq. 36,

$$\begin{aligned}
t_{2} &=\frac{1}{\omega_{0}}\left[\sin ^{-1}\left(\frac{V_{s}}{I_{o} \sqrt{L_{r} / C_{r}}}\right)+\pi\right]+t_{1} \\
&=\frac{1}{4.61\left(10^{6}\right)}\left[\sin ^{-1} \frac{20}{(5)(4.61)}+\pi\right]+0.188 \mu \mathrm{s}=1.10 \mu \mathrm{s}
\end{aligned}$$

From Eq. 39,

$$t_{3}=\left(\frac{L_{r} I_{o}}{V_{s}}\right)\left\{1-\cos \left[\omega_{0}\left(t_{2}-t_{1}\right)\right]\right\}+t_{2}$$

$$=\left(\frac{10^{-6}(5)}{20}\right)\left\{1-\cos \left[(4.61)\left(10^{6}\right)(1.10-0.188)\left(10^{-6}\right)\right]\right\}+1.10 \mu \mathrm{s}=1.47 \mu \mathrm{s}$$


Equation 44 is used to determine the proper switching frequency,

$$\begin{array}{l}
V_{o}=V_{s}\left[1-f_{s}\left(t_{3}-\frac{t_{1}}{2}\right)\right] \\
10=20\left[1-f_{s}\left(1.47-\frac{0.188}{2}\right)\left(10^{-6}\right)\right] \\
f_{s}=363 \mathrm{kHz}
\end{array}$$

(b) Peak reverse voltage across $D_{s}$ is the same as peak capacitor voltage. From Eq. 25,

$$V_{D_{\mathrm{rppak}}}=V_{C_{\mathrm{p} \text { pakk }}}=V_{o}+I_{o} \sqrt{\frac{L_{r}}{C_{r}}}=20+(5)(4.61)=33 \mathrm{v}$$
