# Power Computations

Power computations are essential in analyzing and designing power electronics circuits. Basic power concepts are reviewed in this section, with particular emphasis on power calculations for circuits with nonsinusoidal voltages and currents. Extra treatment is given to some special cases that are encountered frequently in power electronics.

## Power and Energy

### Instantaneous Power

The instantaneous power for any device is computed from the voltage across it and the current in it. Instantaneous power is:

$$
p(t)=v(t) i(t)
$$   [Eq. 1]

This relationship is valid for any device or circuit. Instantaneous power is generally a time-varying quantity. If the passive sign convention illustrated in Fig. 1a is observed, the device is absorbing power if $p(t)$ is positive at a specified value of time $t$. The device is supplying power if $p(t)$ is negative. Sources frequently have an assumed current direction consistent with supplying power. With the convention of Fig. 1b a positive $p(t)$ indicates the source is supplying power.

### Energy

Energy, or work, is the integral of instantaneous power. Observing the passive sign convention, energy absorbed by a component in the time interval from $t_{1}$ to $t_{2}$ is

$$
W=\int_{t_{1}}^{t_{2}} p(t) d t
$$   [Eq. 2]

If $v(t)$ is in volts and $i(t)$ is in amperes, power has units of watts and energy has units of joules.

### Average Power

Periodic voltage and current functions produce a periodic instantaneous power function. Average power is the time average of $p(t)$ over one or more periods. Average power $P$ is computed from

$$
P=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} p(t) d t=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} v(t) i(t) d t
$$   [Eq. 3]

where $T$ is the period of the power waveform. Combining [Eqs. 3 and 2] power is also computed from energy per period.

$$
P=\frac{W}{T}
$$   [Eq. 4]

Average power is sometimes called real power or active power, especially in ac circuits. The term power usually means average power. The total average power absorbed in a circuit equals the total average power supplied.

A special case that is frequently encountered in power electronics is the power absorbed or supplied by a dc source. Applications include battery-charging circuits and dc power supplies. The average power absorbed by a dc voltage source $v(t)=V_{\mathrm{dc}}$ that has a periodic current $l(t)$ is derived from the basic definition of average power in [Eq. 3]:

$$
P_{\mathrm{dc}}=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} v(t) t(t) d t=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} V_{\mathrm{dc}} i(t) d t
$$

Bringing the constant $V_{\mathrm{dc}}$ outside of the integral gives

$$
P_{\mathrm{dc}}=V_{\mathrm{dc}}\left[\frac{1}{T} \int_{t_{0}}^{t_{0}+T} i(t) d t\right]
$$

The term in brackets is the average of the current waveform. Therefore, average power absorbed by a dc voltage source is the product of the voltage and the average current.

$$
P_{\mathrm{dc}}=V_{\mathrm{dc}} I_{\mathrm{avg}}
$$   [Eq. 5]

Similarly, average power absorbed by a dc source $l(t)=I_{\mathrm{dc}}$ is

$$
P_{\mathrm{dc}}=I_{\mathrm{dc}} V_{\mathrm{avg}}
$$   [Eq. 6]

### Effective Values: RMS

The effective value of a voltage or current is also known as the root-mean-square (rms) value. The effective value of a periodic voltage waveform is based on the average power delivered to a resistor. For a dc voltage across a resistor,

$$
P=\frac{V_{d c}^{2}}{R}
$$   [Eq. 34]

For a periodic voltage across a resistor, effective voltage is defined as the voltage that is as effective as the dc voltage in supplying average power. Effective voltage can be computed using the equation

$$
P=\frac{V_{\mathrm{eff}}^{2}}{R}
$$   [Eq. 35]

Computing average resistor power from [Eq. 3] gives

$$
\begin{aligned}
P &=\frac{1}{T} \int_{0}^{T} p(t) d t=\frac{1}{T} \int_{0}^{T} v(t) i(t) d t=\frac{1}{T} \int_{0}^{T} \frac{v^{2}(t)}{R} d t \\
&=\frac{1}{R}\left[\frac{1}{T} \int_{0}^{T} v^{2}(t) d t\right]
\end{aligned}
$$   [Eq. 36]

Equating the expressions for average power in [Eqs. 35 and 36] gives

$$
P=\frac{V_{\text {eff }}^{2}}{R}=\frac{1}{R}\left[\frac{1}{T} \int_{0}^{T} v^{2}(t) d t\right]
$$

or

$$
V_{\mathrm{eff}}^{2}=\frac{1}{T} \int_{0}^{T} v^{2}(t) d t
$$

resulting in the expression for effective or $\mathrm{rms}$ voltage

$$
V_{\mathrm{eff}}=V_{\mathrm{rms}}=\sqrt{\frac{1}{T} \int_{0}^{T} v^{2}(t) d t}
$$   [Eq. 37]

The effective value is the square root of the mean of the square of the voltage hence the term root mean square.
Similarly, rms current is developed from $P=I_{\mathrm{mns}}^{2}$ as

$$
I_{\mathrm{Fms}}=\sqrt{\frac{1}{T} \int_{0}^{T} i^{2}(t) d t}
$$   [Eq. 38]

The usefulness of the rms value of voltages and currents lies in the computing power absorbed by resistances. Additionally, ac power system voltages and currents are invariably given in rms values. Ratings of devices such as electric motros and transformers are often specified in terms of rms voltage and current.

If a periodic voltage is the sum of two periodic voltage waveforms, $v(t)=v_{1}(t)+v_{2}(t)$, the rms value of $v(t)$ is determined from [Eq. 37] as

$$
V_{\mathrm{Tms}}^{2}=\frac{1}{T} \int_{0}^{T}\left(v_{1}+v_{2}\right)^{2} d t=\frac{1}{T} \int_{0}^{T}\left(v_{1}^{2}+2 v_{1} v_{2}+v_{2}^{2}\right) d t
$$

or

$$
V_{\mathrm{ms}}^{2}=\frac{1}{T} \int_{0}^{T} v_{1}^{2} d t+\frac{1}{T} \int_{0}^{T} 2 v_{1} v_{2} d t+\frac{1}{T} \int_{0}^{T} v_{2}^{2} d t
$$

The term containing the product $v_{1} v_{2}$ in the above equation is zero if the functions $v_{1}$ and $v_{2}$ are orthogonal. A condition that satisfies that requirement occurs when $v_{1}$ and $v_{2}$ are sinusoids of different frequencies. For orthogonal functions,

$$
V_{\mathrm{ms}}^{2}=\frac{1}{T} \int_{0}^{T} v_{1}^{2}(t) d t+\frac{1}{T} \int_{0}^{T} v_{2}^{2}(t) d t
$$

Noting that

$$
\frac{1}{T} \int_{0}^{T} v_{1}^{2}(t) d t=V_{1, \text { rms }}^{2} \text { and } \frac{1}{T} \int_{0}^{T} v_{2}^{2}(t) d t=V_{2, \mathrm{mns}}^{2}
$$

then

$$
V_{\mathrm{ms}}=\sqrt{V_{1, \mathrm{ms}}^{2}+V_{2, \mathrm{ms}}^{2}}
$$

If a voltage is the sum of more than two periodic voltages, all orthogonal, the rms value is

$$
V_{\mathrm{mls}}=\sqrt{V_{1, \mathrm{rms}}^{2}+V_{2, \mathrm{mls}}^{2}+V_{3, \mathrm{rms}}^{2}+\ldots}=\sqrt{\sum_{n=1}^{N} V_{n, \mathrm{Fms}}^{2}}
$$   [Eq. 39]

Similarly,

$$
I_{\mathrm{Fms}}=\sqrt{I_{1, \mathrm{rms}}^{2}+I_{2, \mathrm{rms}}^{2}+I_{3, \mathrm{rms}}^{2}+\ldots}=\sqrt{\sum_{n=1}^{N} I_{\mathrm{n}, \mathrm{rms}}^{2}}
$$   [Eq. 40]





### Apparent Power and Power Factor

__Apparent Power $S$__ is the product of rms voltage and rms current magnitudes and is often used in specifying the rating of power equipment such as transformers and electric motors. Apparent power is expressed as

$$
S=V_{\mathrm{rms}} I_{\mathrm{rms}}
$$   [Eq. 41]

In ac circuits (linear circuits with sinusoidal sources), apparent power is the magnitude of complex power [$P+jQ$].

__Power Factor__ The power factor of a load is defined as the ratio of average power to apparent power:

$$
\mathrm{pf}=\frac{P}{S}=\frac{P}{V_{\mathrm{mls}} I_{\mathrm{mls}}}
$$

In sinusoidal ac circuits, the above calculation results in $\mathrm{pf}=\cos \theta$ where $\theta$ is the phase angle between the voltage and current sinusoids. However, that is a special case and should be used only when both voltage and current are sinusoids. In general, power factor must be computed from [Eq. 42]

### Power Computations for Sinusoidal AC Circuits

In general, voltages and/or currents in power electronics circuits are not sinusoidal. However, a nonsinusoidal periodic waveform can be represented by a Fourier series of sinusoids. It is therefore important to understand thoroughly power computations for the sinusoidal case. The following discussion is a review of power computations for ac circuits.

For linear circuits that have sinusoidal sources, all steady-state voltages and currents are sinusoids. Instantaneous power and average power for ac circuits are computed using [Eqs. 1 and 3] as follows:

For any element in an ac circuit, let

$$
\begin{aligned}
v(t) &=V_{m} \cos (\omega t+\theta) \\
i(t) &=I_{m} \cos (\omega t+\phi)
\end{aligned}
$$  [Eq. 43]

Then instantaneous power is

$$
p(t)=v(t) t(t)=\left[V_{m} \cos (\omega t+\theta)\right]\left[I_{m} \cos (\omega t+\phi)\right]
$$   [Eq. 44]

Using the trigonometric identity gives

$$
(\cos A)(\cos B)=\frac{1}{2}[\cos (A+B)+\cos (A-B)]
$$   [Eq. 45]

$$
p(t)=\left(\frac{V_{m} I_{m}}{2}\right)[\cos (2 \omega t+\theta+\phi)+\cos (\theta-\phi)]
$$   [Eq. 46]

Average power is

$$
P=\frac{1}{T} \int_{0}^{T} p(t) d t=\left(\frac{V_{m} I_{m}}{2}\right) \int_{0}^{T}[\cos (2 \omega t+\theta+\phi)+\cos (\theta-\phi)] d t
$$   [Eq. 47]

The result of the above integration can be obtained by inspection. since the first term in the integration is a cosine function, the integral over one period is zero because of equal areas above and below the time axis. The second term in the integration is the constant $\cos (\theta-\phi),$ which has an average value of $\cos (\theta-\phi)$ Therefore, the average power in any element in an ac circuit is

$$
P=\left(\frac{V_{m} I_{m}}{2}\right) \cos (\theta-\phi)
$$   [Eq. 48]

This equation is frequently expressed as

$$
P=V_{\mathrm{rms}} I_{\mathrm{rms}} \cos (\theta-\phi)
$$   [Eq. 49]

where $V_{\mathrm{mas}}=V_{m} / \sqrt{2}, I_{\mathrm{mas}}=I_{m} / \sqrt{2}$, and $\theta-\phi$ is the phase angle between voltage and current. The power factor is determined to be $\cos (\theta-\phi)$ by using [Eq. 42] In the steady state, no net power is absorbed by an inductor or a capacitor. The term reactive power is commonly used in conjunction with voltages and currents for inductors and capacitors. Reactive power is characterized by energy storage during one-half of the cycle and energy retrieval during the other half. Reactive power is computed with a relationship similar to [Eq. 49]:

$$
Q=V_{\mathrm{rms}} I_{\mathrm{rms}} \sin (\theta-\phi)
$$   [Eq. 50]

By convention, inductors absorb positive reactive power and capacitors absorb negative reactive power. Complex power combines real and reactive powers for ac circuits:

$$
\mathrm{S}=P+j Q=\left(\mathrm{V}_{\mathrm{rms}}\right)\left(\mathrm{I}_{\mathrm{mns}}\right)^{*}
$$   [Eq. 51]

In the above equation, $\mathbf{V}_{\mathrm{mns}}$ and $\mathbf{I}_{\mathrm{mns}}$ are complex quantities often expressed as phasors (magnitude and angle), and $\left(\mathbf{I}_{\mathrm{rms}}\right)^{*}$ is the complex conjugate of phasor current, which gives results consistent with the convention that inductance, or lagging current, absorbs reactive power. Apparent power in ac circuits is the magnitude of complex power:

$$
S=|\mathbf{S}|=\sqrt{P^{2}+Q^{2}}
$$   [Eq. 52]

It is important to note that the complex power in [Eq. 52] and power factor of $\cos (\theta-\phi)$ for sinusoidal ac circuits are special cases and are not applicable to nonsinusoidal voltages and currents.

### Power Computations for Nonsinusoidal Periodic Waveforms

Power electronics circuits typically have voltages and/or currents that are periodic but not sinusoidal. For the general case, the basic definitions for the power terms described at the beginning of this chapter must be applied. Acommon error that is made when doing power computations is to attempt to apply some special relationships for sinusoids to waveforms that are not sinusoids.

The Fourier series can be used to describe nonsinusoidal periodic waveforms
in terms of a series of sinusoids. The power relationships for these circuits can be
expressed in terms of the components of the Fourier series.



__Fourier Series__

A nonsinusoidal periodic waveform that meets certain conditions can be described by a Fourier series of sinusoids. The Fourier series for a periodic function $f(t)$ can be expressed in trigonometric form as

$$
f(t)=a_{0}+\sum_{n=1}^{\infty}\left[a_{n} \cos \left(n \omega_{0} t\right)+b_{n} \sin \left(n \omega_{0} t\right)\right]
$$   [Eq. 53]

where

$$
a_{0}=\frac{1}{T} \int_{-T / 2}^{T / 2} f(t) d t
$$   [Eq. 54a]

$$
a_{n}=\frac{2}{T} \int_{-T / 2}^{T / 2} f(t) \cos \left(n \omega_{0} t\right) d t
$$   [Eq. 54b]

$$
b_{n}=\frac{2}{T} \int_{-T / 2}^{T / 2} f(t) \sin \left(n \omega_{0} t\right) d t
$$   [Eq. 54c]

sines and cosines of the same frequency can be combined into one sinusoid, resulting in an alternative expression for a Fourier series:

$$
f(t)=a_{0}+\sum_{n=1}^{\infty} C_{n} \cos \left(n \omega_{0} t+\theta_{n}\right)
$$   [Eq. 55a]

where

$$
C_{n}=\sqrt{a_{n}^{2}+b_{n}^{2}} \quad \text { and } \quad \theta_{n}=\tan ^{-1}\left(\frac{-b_{n}}{a_{n}}\right)
$$   [Eq. 55b]

or

$$
f(t)=a_{0}+\sum_{n=1}^{\infty} C_{n} \sin \left(n \omega_{0} t+\theta_{n}\right)
$$   [Eq. 56a]

where

$$
C_{n}=\sqrt{a_{n}^{2}+b_{n}^{2}} \quad \text { and } \quad \theta_{n}=\tan ^{-1}\left(\frac{a_{n}}{b_{n}}\right)
$$   [Eq. 56b]

The term $a_{0}$ is a constant that is the average value of $f(t)$ and represents a dc voltage or current in electrical applications. The coefficient $C_{1}$ is the amplitude of the term at the fundamental frequency $\omega_{0}$. Coefficients $C_{2}, C_{3}, \ldots$ are the amplitudes of the harmonics that have frequencies $2 \omega_{0}, 3 \omega_{0}, \ldots$

The rms value of $f(t)$ can be computed from the Fourier series:

$$
F_{\mathrm{mls}}=\sqrt{\sum_{n=0}^{\infty} F_{n, \mathrm{ms}}^{2}}=\sqrt{a_{0}^{2}+\sum_{n=1}^{\infty}\left(\frac{C_{n}}{\sqrt{2}}\right)^{2}}
$$



__Average Power__

If periodic voltage and current waveforms represented by the Fourier series

$$
\begin{array}{l}
v(t)=V_{0}+\sum_{n=1}^{\infty} V_{n} \cos \left(n \omega_{0} t+\theta_{n}\right) \\
l(t)=I_{0}+\sum_{n=1}^{\infty} I_{n} \cos \left(n \omega_{0} t+\phi_{n}\right)
\end{array}
$$   [Eq. 58]

exist for a device or circuit, then average power is computed from [Eq. 3].

$$
P=\frac{1}{T} \int_{0}^{T} v(t) i(t) d t
$$

The average of the products of the de terms is $V_{0} I_{0} .$ The average of voltage and current products at the same frequency is described by [Eq. 49], and the average of voltage and current products of different frequencies is zero. Consequently, average power for nonsinusoidal periodic voltage and current waveforms is

$$
\begin{array}{l}
P=\sum_{n=0}^{\infty} P_{n}=V_{0} I_{0}+\sum_{n=1}^{\infty} V_{n, \operatorname{mis}} I_{n, \mathrm{Fms}} \cos \left(\theta_{n}-\phi_{n}\right) \\
\quad \text { or } \\
\quad P=V_{0} I_{0}+\sum_{n=1}^{\infty}\left(\frac{V_{n, \max } I_{n, \max }}{2}\right) \cos \left(\theta_{n}-\phi_{n}\right)
\end{array}
$$   [Eq. 59]

Note that total average power is the sum of the powers at the frequencies in the Fourier series.



__Nonsinusoidal Source and Linear Load__

If a nonsinusoidal periodic voltage is applied to a load that is a combination of linear elements, the power absorbed by the load can be determined by using superposition. A nonsinusoidal periodic voltage is equivalent to the series combination of the Fourier series voltages, as illustrated in Fig. 10. The current in the load can be determined using superposition, and [Eq. 59] can be applied
to compute average power. Recall that superposition for power is not valid when the sources are of the same frequency.

__Sinusoidal Source and Nonlinear Load__
If a sinusoidal voltage source is applied to a nonlinear load, the current waveform will not be sinusoidal but can be represented as a Fourier series. If voltage is the sinusoid

$$
v(t)=V_{1} \sin \left(\omega_{0} t+\theta_{1}\right)
$$   [Eq. 60]

and current is represented by the Fourier series

$$
i(t)=I_{0}+\sum_{n=1}^{\infty} I_{n} \sin \left(n \omega_{0} t+\phi_{n}\right)
$$   [Eq. 61]

then average power absorbed by the load (or supplied by the source) is computed from [Eq. 59] as

$$
\begin{aligned}
P &=V_{0} I_{0}+\sum_{n=1}^{\infty}\left(\frac{V_{n,} \max I_{n, \max }}{2}\right) \cos \left(\theta_{n}-\phi_{n}\right) \\
&=(0)\left(I_{0}\right)+\left(\frac{V_{1} I_{1}}{2}\right) \cos \left(\theta_{1}-\phi_{1}\right)+\sum_{n=2}^{\infty} \frac{(0)\left(I_{n, \max }\right)}{2} \cos \left(\theta_{n}-\phi_{n}\right) \quad(2-62) \\
&=\left(\frac{V_{1} I_{1}}{2}\right) \cos \left(\theta_{1}-\phi_{1}\right)=V_{1, \mathrm{mIs}} I_{1, \mathrm{rms}} \cos \left(\theta_{1}-\phi_{1}\right)
\end{aligned}
$$   [Eqs. 62]

_Note that the only nonzero power term is at the frequency of the applied voltage._ The power factor of the load is computed from [Eq. 42]

$$
\begin{array}{l}
\mathrm{pf}=\frac{P}{S}=\frac{P}{V_{\mathrm{mns}} I_{\mathrm{mns}}} \\
\mathrm{pf}=\frac{V_{1, \mathrm{rms}} I_{1, \mathrm{rms}} \cos \left(\theta_{1}-\phi_{1}\right)}{V_{1, \mathrm{mns}} I_{\mathrm{ms}}}=\left(\frac{I_{1, \mathrm{ms}}}{I_{\mathrm{rms}}}\right) \cos \left(\theta_{1}-\phi_{1}\right)
\end{array}
$$   [Eqs. 63]

where $\mathrm{rms}$ current is computed from

$$
I_{\mathrm{rms}}=\sqrt{\sum_{n=0}^{\infty} I_{n, \mathrm{mns}}^{2}}=\sqrt{I_{0}^{2}+\sum_{n=1}^{\infty}\left(\frac{I_{n}}{\sqrt{2}}\right)^{2}}
$$   [Eqs. 64]

Note also that for a sinusoidal voltage and a sinusoidal current, $\mathrm{pf}=\cos \left(\theta_{1}-\phi_{1}\right),$ which is the power factor term commonly used in linear circuits and is called ___The Displacement Power Factor___. The ratio of the $\mathrm{rms}$ value of the fundamental frequency to the total $\mathrm{rms}$ value, $I_{1, \operatorname{mas}} / I_{\max }$ in Eq. $(2-63),$ is ___The Distortion Factor (DF)___.

$$
\mathrm{DF}=\frac{I_{1, \mathrm{rms}}}{I_{\mathrm{ms}}}
$$   [Eqs. 65]

The distortion factor represents the reduction in power factor due to the nonsinusoidal property of the current. Power factor is also expressed as

$$
\mathrm{pf}=\left[\cos \left(\theta_{1}-\phi_{1}\right)\right] \mathrm{DF}
$$   [Eqs. 66]

___Total Harmonic Distortion (THD)___ is another term used to quantify the nonsinusoidal property of a waveform. THD is the ratio of the rms value of all the nonfundamental frequency terms to the rms value of the fundamental frequency term.

$$
\mathrm{THD}=\sqrt{\frac{\sum I_{n, \mathrm{ms}}^{2}}{I_{1, \mathrm{ms}}^{2}}}=\frac{\sqrt{\sum I_{n, \mathrm{mns}}^{2}}}{I_{1, \mathrm{ms}}}
$$   [Eqs. 67]

THD is equivalently expressed as

$$
\mathrm{THD}=\sqrt{\frac{I_{\mathrm{mis}}^{2}-I_{1, \mathrm{ms}}^{2}}{I_{1, \mathrm{mls}}^{2}}}
$$   [Eqs. 68]

Total harmonic distortion is often applied in situations where the de term is zero, in which case THD may be expressed as

$$
\mathrm{THD}=\frac{\sqrt{\sum_{n=2}^{\infty} I_{n}^{2}}}{I_{1}}
$$   [Eqs. 69]

Another way to express the distortion factor is

$$
\mathrm{DF}=\sqrt{\frac{1}{1+(\mathrm{THD})^{2}}}
$$   [Eqs. 70]

Reactive power for a sinusoidal voltage and a nonsinusoidal current can be expressed as in [Eq. 50]. The only nonzero term for reactive power is at the voltage frequency:

$$
Q=\frac{V_{1} I_{1}}{2} \sin \left(\theta_{1}-\phi_{1}\right)
$$   [Eqs. 71]

With $P$ and $Q$ defined for the nonsinusoidal case, apparent power $S$ must include a term to account for the current at frequencies which are different from the voltage frequency. The term distortion volt-amps $D$ is traditionally used in the computation of $S$,

$$
S=\sqrt{P^{2}+Q^{2}+D^{2}}
$$   [Eqs. 72]

where

$$
D=V_{1,}, \mathrm{ms} \sqrt{\sum_{n \neq 1}^{\infty} I_{n, \mathrm{Fms}}^{2}}=\frac{V_{1}}{2} \sqrt{\sum_{n \neq 1}^{\infty} I_{n}}
$$   [Eqs. 73]

Other terms that are sometimes used for nonsinusoidal current (or voltages) are ___form factor___ and ___crest factor.___

$$
\text { Form factor }=\frac{I_{\mathrm{mns}}}{I_{\mathrm{avg}}}
$$   [Eqs. 74]

$$
\text { Crest factor }=\frac{I_{\text {peak }}}{I_{\text {mis }}}
$$   [Eqs. 75]


# Summary

* Instantaneous power is the product of voltage and current at a particular time:

$$
p(t)=v(t) i(t)
$$

Using the passive sign convention, the device is absorbing power if $p(t)$ is positive, and the device is supplying power if $p(t)$ is negative.

* Power usually refers to average power, which is the time average of periodic instantaneous power:

$$
P=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} v(t) i(t) d t=\frac{1}{T} \int_{t_{0}}^{t_{0}+T} p(t) d t
$$

* The rms value is the root-mean-square or effective value of a voltage or current waveform.

$$
\begin{aligned}
V_{\mathrm{ms}} &=\sqrt{\frac{1}{T} \int v^{2}(t) d t} \\
I_{\mathrm{rms}} &=\sqrt{\frac{1}{T} \int_{0}^{T} i^{2}(t) d t}
\end{aligned}
$$

* Apparent power is the product of rms voltage and current.

$$
S=V_{\mathrm{mis}} I_{\mathrm{rms}}
$$

* Power factor is the ratio of average power to apparent power.

$$
\mathrm{pf}=\frac{P}{S}=\frac{P}{V_{\mathrm{ms}} I_{\mathrm{rms}}}
$$


* For nonsinusoidal periodic waveforms, average power may be computed from the basic definition, or the Fourier series method may be used. The Fourier series method treats each frequency in the series separately and uses superposition to compute total power.

$$
P=\sum_{n=0}^{\infty} P_{n}=V_{0} I_{0}+\sum_{n=1}^{\infty} V_{n, \operatorname{mos}} I_{n, \min } \cos \left(\theta_{n}-\phi_{n}\right)
$$

* For periodic currents, the average voltage across an inductor is zero.

* For periodic voltages, the average current in a capacitor is zero. 

* For inductors and capacitors that have periodic voltages and currents, the average power is zero. Instantaneous power is generally not zero because the device stores energy and then returns energy to the circuit.

* A simulation using the program PSpice may be used to obtain not only voltage and current waveforms but also instantaneous power, energy, rms values, and average power by using the numerical capabilities of the graphic postprocessor program
