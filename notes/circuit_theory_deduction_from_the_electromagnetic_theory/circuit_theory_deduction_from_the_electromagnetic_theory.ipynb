# Circuit Theory Deduction from The Electromagnetic Theory

## Introduction

Circuit Theory is a fundamental component of the electrical and electronic engineering career curricula. The validity of this theory, however, is subjected to the fulfillment of certain physical premises that are often ignored in basic circuit and electromagnetism courses. In this notes it will be deduced the main equations of the Circuit Theory from the Electromagnetic Theory with the purpose of facilitating the understanding of its limitations and of providing a more precise vision of its physical meaning.

## 1 La Teoría de Circuitos

La Teoría de Circuitos se aplica en el análisis de circuitos eléctricos de alta tensión, motores eléctricos, circuitos electrónicos analógicos y digitales de baja frecuencia, circuitos electrónicos de potencia, etc. Un circuito eléctrico consiste en general de un conjunto de elementos concentrados de los tipos: resistor, capacitor, inductor, memristor, generador de voltaje y generador de corriente, conectados entre si de acuerdo a una topología arbitraria.

La Teoría de Circuitos se fundamenta en tres leyes fundamentales, a saber: la Ley de OnM, la Ley de Voltaje de KIRCHHOFF y la Ley de Corriente de KIRCHHOFF (Boylestad, 2008; Sedra & Smith, 2017) . La Ley de OHM establece que la corriente I que circula a través de una resistencia R es proporcional al voltaje $\mathrm{V}$ aplicado a $\mathrm{R}$ :

$$
I=\frac{V}{R}
$$   [Eq. 1]

Las leyes de voltaje y de corriente de KIRCHIIOFF establecen que la suma algebraica de voltajes a lo largo de una malla de un circuito eléctrico [Fig. 1] la suma algebraica de corrientes en un nodo [Fig. 2], respectivamente, son ambas nulas:

$$
\sum_{n} v_{n}=v_{g}-v_{L}-v_{R}-v_{C}=0
$$   [Eq. 2]
$$
\sum_{n} i_{n}=i_{g}-i_{L}-i_{R}-i_{C}=0
$$   [Eq. 3]

donde $v_{n}$ es el voltaje a lo largo del componente n-ésimo de la malla, e $i_{n}$ es la corriente que accede al nodo desde la rama n-ésima, respectivamente.

[En estas ecuaciones se fundamenta más del $60 \%$ de las asignaturas que forman parte
del currículo de ingeniería eléctrica y electrónica.]

## 2. La Teoría Electromagnética
La Teoría Electromagnética es uno de los cuatro grandes cuerpos de la física que se considera completon y describe todos los fenómenos electromagnéticos macroscópicos de la naturaleza incluyendo, en particular, los circuitos eléctricos. La Teoría Electromagnética se resume en las cuatro Ecuaciones de MAXWELL:

$$
\nabla \times \mathbf{E}=-\frac{\partial \mathbf{B}}{\partial \mathrm{t}}
$$ [Eq.4]

$$
\nabla \times \mathbf{H}=\mathbf{J}+\frac{\partial \mathbf{D}}{\partial \mathrm{t}}
$$ [Eq.5]

$$
\nabla \cdot \mathbf{D}=\rho_{\mathbf{v}}
$$ [Eq.6]

$$
\nabla \cdot \mathbf{B} =0
$$ [Eq.7]


donde $\mathrm{E}[\mathrm{V} / \mathrm{m}]$ es la intensidad del campo eléctrico, $\mathrm{H}[\mathrm{A} / \mathrm{m}]$ es la intensidad de campo magnético, $\mathrm{D}[\mathrm{C} / \mathrm{m} 2]$ es la densidad de flujo eléctrico, $\mathrm{B}[\mathrm{T}]$ es la densidad de flujo, magnético, $\rho_{v}[\mathrm{C} / \mathrm{m} 3]$ es la densidad volumétrica de cargas libres $\mathrm{y} \mathrm{J}[\mathrm{A} / \mathrm{m} 2]$ es la densidad de corriente libre [Las unidades de medida pertenecen al sistema internacional de unidades (SI)].

La Ecuación de Inducción de FARADAY $\nabla \times \mathrm{E}=-\frac{\partial \mathrm{B}}{\partial \mathrm{t}}$ establece que todo campo magnético variante en el tiempo "engendra" un campo eléctrico solenodial que rota transversalmente alrededor de aquel.

<br><br/>
<br><br/>

La Ecuación de AMPERE-MAXWELL $\nabla \times \mathrm{H}=\mathrm{J}+\frac{\partial \mathrm{D}}{\partial \mathrm{t}}$ establece que todo campo eléctrico variante en el tiempo y toda corriente eléctrica "engendran" un campo magnético solenodial que rota transversalmente alrededor de los primeros. 

<br><br/>

<center> Figure 3.$\quad$ Illustration of Ampere's law. </center>
<br><br/>
<br><br/>

La Ecuación de GAUSS $\nabla \cdot \mathrm{D}=\rho_{v}$ establece que las cargas eléctricas producen un campo eléctrico que diverge, o converge, según su polaridad, tal que las cargas positivas constituyen "fuentes" positivas del campo, y las negativas "sumideros". La ecuación $\nabla \cdot \mathrm{B}=0,$ a veces denominada Ecuación de Gauss para el campo magnético, establece que este no posee fuentes escalares.

Las Ecuaciones de MAXWELL se suelen acompañar con las ecuaciones constitutivas de la materia:

$$
\mathrm{D}=\varepsilon \mathrm{E}
$$ [Eq. 8]

$$
\mathrm{B}=\mu \mathrm{H}
$$ [Eq. 9]

$$
\mathrm{J}=\sigma \mathrm{E}
$$ [Eq. 10]

donde $\varepsilon[\mathrm{F} / \mathrm{m}]$ es la permitividad eléctrica del medio, $\mu[\mathrm{H} / \mathrm{m}]$ es la permeabilidad magnética del medio y $\sigma[\mathrm{S} / \mathrm{m}]$ es la conductividad del medio.

Las [Eqs. 8 - 10] esta última, por cierto, conocida como Ley de OHM, modelan la materia como si de un sistema lineal se tratase, donde los parámetros intrínsecos $\varepsilon, \mu$ y $\sigma$ vendrían a ser las funciones de transferencia, los campos a la derecha las entradas (funciones forzantes), y los campos a la izquierda las salidas (respuestas del sistema).

## 3 Deducción de la Teoría de Circuitos a partir de la Teoría Electromagnética

La Teoría de Circuitos, como se vera en seguida, es un subconjunto de la Teoría Electromagnética.



### 3.1 Deducción de la Ecuación de Voltaje de KIRCHHOFF

El campo eléctrico de la Ec. (4) es de naturaleza distinta a la del campo eléctrico de la Ec. (6). El primero es solenoidal y el segundo irrotacional. La Ecuación de Voltaje de KIRCHHOFF presupone que el campo eléctrico sea del segundo tipo: $\nabla \times \mathrm{E}=0 .$ Procedamos sin embargo considerando, como ocurre en general en el mundo real, que esto no es verdad, salvo en el caso $\mathrm{DC}$, usando la versión integral de la Ec. (4):

$$
\oint_{\Gamma} \mathrm{E} \cdot \mathrm{d} \ell=\int_{\mathrm{S}(\Gamma)}-\frac{\partial \mathrm{B}}{\partial \mathrm{t}} \cdot \mathrm{ds}
$$ [Eq. 11]

la cual aplicaremos a la malla de la Fig.1(a), tomando en cuenta que $v=-\int_{-}^{+} \mathrm{E} \cdot \mathrm{d} \ell$. De esta forma se obtiene:

$$
\sum_{n} v_{n}=\int_{S(\Gamma)}-\frac{\partial B}{\partial t} \cdot d s
$$ [Eq. 12]

$$
v_{g}-v_{L}-v_{R}-v_{C}=\int_{S(\Gamma)}-\frac{\partial \mathbf{B}}{\partial t} \cdot d s
$$ [Eq. 13]

la cual se "parece , mucho a la Ecuación de Voltaje de KIRCHHOFF [Eq. 2] salvo por el término el término  $\int_{S(\Gamma)}-\frac{\partial B}{\partial t} \cdot d s$ a la derecha y asumiendo que el circuito no irradie . ¿Cómo se interpreta este término y en que medida perturba la validez de la Ecuación de Voltaje de KIRCHHOFF?

Los voltajes a la izquierda de la Ec. (13) son fácilmente reconocibles a partir del diagrama esquemático del circuito de la [Fig. 1a]. El término a la derecha, por otro lado, representa la f.e.m. inducida en el circuito por el campo magnético de la corriente eléctrica que recorre la malla. Este término bien podría modelarse como un inductor aparásito" cuyo valor, como puede verse de la Ec. (13), depende de la frecuencia de operación del circuito a través de la derivada $\partial / \partial t$ del campo magnético y del tamaño del propio circuito a través de su superficie $S(\Gamma)$ en la integral de flujo. Solo si la variación de voltaje por unidad de tiempo del generador es suficientemente pequeña, o si el tamaño del circuito es relativamente pequeño, se podrá escribir:

$$
v_{g}-v_{L}-v_{R}-v_{C} \approx 0
$$



### 3.2 Deducción de la Ecuación de Corriente de KIRCHHOFF

La Ecuación de Corriente de KIRCHHOFF se deriva de la ecuación de rotacional del campo
magnético (5). En efecto, considerando que un campo rotacional no diverge (McQuistan,
1965), al tomar la divergencia de la Ec. (5) y tomando en cuenta la Ec. (6), se obtiene la Ecuación de Continuidad de la Corriente:

$$
\nabla \cdot \nabla \times \mathbf{H}=\nabla \cdot\left(\mathbf{J}+\frac{\partial \mathbf{D}}{\partial t}\right)
$$   [Eq. 15]

$$
0=\nabla \cdot \mathbf{J}+\nabla \cdot\left(\frac{\partial \mathbf{D}}{\partial \mathrm{t}}\right)
$$   [Eq. 16]

$$
\nabla \cdot \mathbf{J}=-\frac{\partial}{\partial \mathrm{t}} \nabla \cdot \mathbf{D}
$$   [Eq. 17]

$$
\nabla \cdot \mathbf{J}=-\frac{\partial \rho_{v}}{\partial t}
$$   [Eq. 18]

Al aplicar la versión integral de la Ec. (18) al nodo del circuito 1(b), tomando en cuenta que $i=\int_{\Delta \mathrm{S}} \mathrm{J} \cdot \mathrm{ds},$ se obtiene:

$$
\oint_{\mathrm{S}} \mathbf{J} \cdot \mathbf{d s}=\int_{V(\mathrm{S})}-\frac{\partial \rho_{v}}{\partial \mathrm{t}} \mathrm{d} v
$$   [Eq. 19]

$$
\sum_{n} i_{n}=\int_{V(S)}-\frac{\partial \rho_{v}}{\partial t} d v
$$   [Eq. 20]

$$
i_{g}-i_{R}-i_{L}-i_{C}=\int_{V(S)}-\frac{\partial \rho_{v}}{\partial t} d v
$$   [Eq. 21]

ecuación que se "parece" mucho a la Ecuación de Corriente de KIRCHHOFF [Eq. 3] salvo por el término $\int_{V(S)}-\frac{\partial \rho_{v}}{\partial t} d v$ a la derecha. Cabe preguntarse igualmente ?Cómo se interpreta este término y en que medida perturba la validez de la Ecuación de Corriente de KIRCHHOFF?

La respuesta es similar. Las corrientes a la izquierda de la [Eq. 21] son fácilmente reconocibles a partir del diagrama esquemático del circuito de la Fig. 1(b). El término a la derecha, por otro lado, representa el flujo de corriente de desplazamiento desde el nodo o la variación temporal de la densidad de las cargas acumuladas en el nodo. Este término bien podría modelarse como un _capacitor parásito_ cuyo valor, como puede verse de la [Eq. 21], también depende de la frecuencia de operación del circuito a través de la derivada $\partial / \partial t$ y del tamaño del nodo a través de la superficie $S(\Gamma)$ en la integral de flujo. Solo si la variación de voltaje por unidad de tiempo del generador es suficientemente pequeña, o si el tamaño del circuito es relativamente pequeño, se podrá escribir:

$$
i_{g}-i_{L}-i_{R}-i_{C} \approx 0
$$

## 4 Relación de la frecuencia con el tamaño del circuito en la Teoría de Circuitos

Si bien la frecuencia y el tamaño del circuito determinan las condiciones de validez de las ecuaciones de KIRCHHOFF, ambos parámetros no participan independientemente en el establecimiento de estas condiciones, sino todo lo contrario, están íntimamenten relacionados. Para ver este asunto de manera intuitiva pensemos en el generador como un dispositivo que inyecta una señal a al circuito cada medio período $T=1 / f$ donde $f$ es la frecuencia del generador. El generador inyecta la señal

$$
v_{g}(t)=A \cos (2 \pi f t) V
$$

Es razonable admitir que  la señal se transmitan a lo largo del circuito a una velocidad finita. Por tanto a lo largo del circuito los valores de voltaje forzados por el generador se propagarán tomándose un tiempo igualmente finito y distinto para alcanzar cada punto del circuito. Llámese este tiempo tiempo de transito $\mathrm{t}_{\mathrm{T}}$. Se pueden vislumbrar dos casos extremos:  
* (a) que el tiempo de tránsito sea mucho menor que el período del voltaje del generador $t_{\mathrm{T}} \ll \mathrm{T}$. 
* (b) que el tiempo de tránsito sea mucho mayor que el periodo del voltaje del generador $t_{\top} \gg T$.

Para analizar el impacto de estas dos condiciones asumamos que el circuito bajo análisis alcanza su mayor dimensión $D$ en la dirección del eje $z$ y que tiene la forma simplificada que se muestra en la Fig. 2.

El voltaje, al propagarse en el circuito deberá responder entonces a una ecuación de la forma:

$$
v(t, z)=\alpha v_{g}\left(t-z / v_{p}\right)=\alpha A \cos \left[2 \pi f\left(t-z / v_{p}\right)\right]
$$   [Eq. 23]

donde $v_{p}$ es la velocidad de propagación del voltaje, aproximadamente $v_{p} \approx c / \sqrt{\epsilon_{r}}$ donde $c=3 \times 10^{8} \mathrm{m} / \mathrm{s}$ es la velocidad de la luz en el vacío, $\epsilon_{\mathrm{r}}$ es la constante dieléctrica del medio en el que se encuentra inmerso el circuito, $z / v_{\mathrm{p}}$ es el tiempo de tránsito hasta el punto $z$ del circuito $y \propto$ modela la partición del voltaje en el punto de conexión del generador con el circuito.

<center> Figure 2.$\quad$ Modelo simplificado del circuito visto como linea de transmisión </center>
<br><br/>

El término $z / v_{\mathrm{p}}$ corresponde al retardo acumulado por el voltaje hasta alcanzar el punto $z$ del circuito. Este retardo será mayor para los puntos más alejados que para los más cercanos. El término $z / v_{\mathrm{p}}$ es también un termino de fase que se puede poner en función de la longitud de onda $\lambda$. La longitud de onda es un perfodo espacial que se establece como consecuencia de la propagación del voltaje del generador en el circuito. Esta fase se puede reescribir como $z / v_{\mathrm{p}}=\mathrm{\kappa} z / \omega,$ donde $\omega=2 \pi \mathrm{f}$ es la frecuencia angular $\mathrm{y} \mathrm{k}=2 \pi / \lambda$ es la constante de propagación. Asi, la [Eq. 23] se puede reescribir de la forma:

$$
v(t, z)=\alpha A \cos (\omega t-\kappa z)
$$

Con la [Eq. 24] se pueden analizar ahora los dos casos planteados.



### 4.1 Caso $t_{T} \gg T$

La condición $t_{\top} \gg T$ es equivalente a la condicion $D \gg \lambda,$ lo cual implica que se acumularán grandes retardos, o grandes diferencias de fase, a lo largo del circuito. Podría ocurrir, por ejemplo, para dos puntos distintos del circuito, dígase $z_{1}$ y $z_{2}$, conectados simplemente mediante un conductor, lo que comúnmente suele llamarse un _corto circuito_ [Fig. 2] que $\mathrm{kz}_{1}-\mathrm{k} z_{2}=\pi$ dando lugar a un voltaje no nulo entre estos dos puntos. Y sin embargo la Teoría de Circuitos no ofrece ninguna explicación para este resultado.



### 4.2 Caso $t_{\top} \ll T$

La condición $t_{\top} \ll$ T es equivalente a la condición $D \ll \lambda$. En este caso no se establece entre los distintos puntos del circuito una diferencia de fase significativa, las señales inyectadas por el generador llegan a todos los puntos del circuito prácticamente al mismo tiempo. Ni la corriente ni el voltaje cambian de polaridad sobre ningún tramo conductor del circuito. En este caso la Teoría de Circuitos funciona plenamente y lo explicaría todo.

Con base en lo anterior, conocida la máxima dimensión D de un circuito, cabe preguntarse ¿cuál es la máxima frecuencia, o equivalentemente, cuál es la mínima longitud de onda, que un circuito "tolera" para seguirse comportando como tal? O dada la frecuencia de operación ¿cuál es el tamaño máximo del circuito para que las ecuaciones de KIRCHHOFF predigan con exactitud los voltajes y corrientes en el circuito?

La respuesta cualitativa es $\mathrm{D} \ll \lambda$. Una respuesta cuantitativa que suele usarse en el campo de la integridad de señal y de la compatibilidad electromagnética es $D \leqslant \lambda / 20$ (Schmitt, 2002).
