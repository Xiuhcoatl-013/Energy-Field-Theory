## Boost Converter with Discontinuous Current


The boost converter will also operate for discontinuous inductor current. In some cases, the discontinuous-current mode is desirable for control reasons in the case of a regulated output.

The inductor and diode currents for discontinuous current have the basic waveforms as shown in the next Fig.

<br><br/>
<p align="center"><img src="images/Discontinuous_Inductor_Current_Boost_Converter.png"></p>
<center> Figure: The Inductor Discontinuous Current in the Boost Converter. </center>
<br><br/>


The Boost Converter working with the switch open and the diode off, the inductor current is zero. The average voltage across the inductor is given by:

<br><br/>
$$V_{s} D T+\left(V_{s}-V_{o}\right) D_{1} T=0$$
<br><br/>



which results in


$$V_{o}=V_{s}\left(\frac{D+D_{1}}{D_{1}}\right)$$




The next Figure shows the diode current:

<br><br/>
<p align="center"><img src="images/Discontinuous_Diode_Current_Boost_Converter.png"></p>
<center> Figure: Discontinuous Current in the Boost Converter. The Inductor Current. </center>
<br><br/>

The average diode current is given by:

<br><br/>
$$I_{D}=\frac{1}{T}\left(\frac{1}{2} I_{\max } D_{1} T\right)=\frac{1}{2} I_{\max } D_{1}$$
<br><br/>



Current $I_{max}$ is the same as the change in inductor current when the switch is
closed.

Note that current $I_{max}$ is the same as the change in inductor current when the switch is
closed. that is:

<br><br/>
$$I_{\max }=\Delta i_{L}=\frac{V_{s} D T}{L}$$
<br><br/>



Substituting for $I_{max}$ in prefvious Eq. and setting the result equal to the load current:

<br><br/>
$$I_{D}=\frac{1}{2}\left(\frac{V_{s} D T}{L}\right) D_{1}=\frac{V_{o}}{R}$$
<br><br/>



Then solving for $D_{1}$

<br><br/>
$$D_{1}=\left(\frac{V_{o}}{V_{s}}\right)\left(\frac{2 L}{R D T}\right)$$
<br><br/>



Substituting the preceding expression for D1 into previous Equations, results in the next quadratic equation:

<br><br/>
$$\left(\frac{V_{o}}{V_{s}}\right)^{2}-\frac{V_{o}}{V_{s}}-\frac{D^{2} R T}{2 L}=0$$
<br><br/>



Finally Solving for $V_{\mathrm{o}} / V_{\mathrm{s}}$, it gives:

<br><br/>
$$\frac{V_{o}}{V_{s}}=\frac{1}{2}\left(1+\sqrt{1+\frac{2 D^{2} R T}{L}}\right)$$
<br><br/>



The boundary between continuous and discontinuous current occurs when $D_1 = 1 - D$. Another condition at the boundary is when $I_{min}$ for Inductor Current is zero. Whether the boost converter is operating in the continuous or discontinuous
mode depends on the combination of circuit parameters, including the duty ratio. As the duty ratio for a given boost converter is varied, the converter may go into and out of the discontinuous mode.

## Inductor Resistance

Inductors should be designed to have small resistance to minimize power loss and maximize efficiency. The existence of a small inductor resistance does not substantially change the analysis of other converters like the buck as example. However, inductor resistance princiopally affects performance of the boost converter, and especially at high duty ratios.

For the boost converter, the transfer function is:

<br><br/>
$$\frac{V_{o}}{V_{s}}=\frac{1}{1-D}$$
<br><br/>



To understand the effect of inductor resistance on the output voltage, assume that the inductor current is approximately constant. The source current is the same as the inductor current, and average diode current is the same as average load current. The power supplied by the source must be the same as the power absorbed by the load and the inductor resistance, neglecting other losses.

<br><br/>
$$P_{s}=P_{o}+P_{r_{L}}$$
<br><br/>
$$V_{s} I_{L}=V_{o} I_{D}+I_{L}^{2} r_{L}$$
<br><br/>



where $r_{L}$ is the series resistance of the inductor. The diode current is equal to the inductor current when the switch is off and is zero when the switch is on. Therefore, the average diode current is given by: 

<br><br/>
$$I_{D}=I_{L}(1-D)$$
<br><br/>



Substituting for $I_{D}$ into last Eq. and solving for $V_{s}$ gives:

<br><br/>
$$V_{s}=V_{o}(1-D)+I_{L} r_{L}$$
<br><br/>



Now considering $I_{L}$ in terms of Vo:

<br><br/>
$$I_{L}=\frac{I_{D}}{1-D}=\frac{V_{o} / R}{1-D}$$
<br><br/>



Substituting for $I_{L}$ into last equations:

<br><br/>
$$V_{s}=\frac{V_{o} r_{L}}{R(1-D)}+V_{o}(1-D)$$
<br><br/>



From the las equation, the Transfer function is given by:

<br><br/>
$$\frac{V_{o}}{V_{s}}=\left(\frac{1}{1-D}\right)\left(\frac{1}{1+r_{L} /\left[R(1-D)^{2}\right]}\right)$$
<br><br/>



The preceding equation is similar to that for an ideal converter but includes a correction factor to account for the inductor resistance. The next Figure shows the output voltage of the boost converter with and without inductor resistance.

<br><br/>
<p align="center"><img src="images/Boost_Converter_for_a_Nonideal_Inductor.png"></p>
<center> Figure: Boost Converter for a Nonideal Inductor </center>
<br><br/>



The inductor resistance also has an effect on the power efficiency of converters. Efficiency is the ratio of output power to output power plus losses. For the boost converter:

<br><br/>
$$\eta=\frac{P_{o}}{P_{o}+P_{\text { loss }}}$$
<br><br/>
$$\eta=\frac{V_{o}^{2} / R}{V_{o}^{2} / R+\left(V_{o} / R\right)^{2} /(1-D) r_{L}}=\frac{1}{1+r_{L}\left[R(1-D)^{2}\right]}$$
<br><br/>
