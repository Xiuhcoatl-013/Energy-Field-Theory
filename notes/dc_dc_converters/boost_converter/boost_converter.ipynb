The boost converter is shown in Fig. 06007. This is another switching converter that operates by periodically opening and closing an electronic switch. It is called a boost converter because the output voltage is larger than the input.

__Voltage and Current Relationships__
The analysis assumes the following:

   1. Steady-state conditions exist.
   2. The switching period is $T$, and the switch is closed for time $D T$ and open for $(1-D) T$
   3. The inductor current is continuous (always positive).
   4. The capacitor is very large, and the output voltage is held constant at voltage $V_{o}$
   5. The components are ideal.

The analysis proceeds by examining the inductor voltage and current for the switch closed and again for the switch open.

__Analysis for the Switch Closed__ When the switch is closed, the diode is reversebiased. Kirchhoff's voltage law around the path containing the source, inductor, and closed switch is

$$
v_{L}=V_{s}=L \frac{d i_{L}}{d t} \quad \text { or } \quad \frac{d i_{L}}{d t}=\frac{V_{s}}{L}
$$ [Eq. 06024]

The rate of change of current is a constant, so the current increases linearly while the switch is closed, as shown in Fig. 06009b. The change in inductor current is computed from

$$
\frac{\Delta i_{L}}{\Delta t}=\frac{\Delta i_{L}}{D T}=\frac{V_{s}}{L}
$$

Solving for $\Delta i_{L}$ for the switch closed,

$$
\left(\Delta i_{L}\right)_{\text {closed }}=\frac{V_{s} D T}{L}
$$ [Eq. 06025]

__Analysis for the Switch Open__ When the switch is opened, the inductor current cannot change instantaneously, so the diode becomes forward-biased to provide a path for inductor current. Assuming that the output voltage $V_{o}$ is a constant, the voltage across the inductor is

$$
\begin{array}{c}
v_{L}=V_{s}-V_{o}=L \frac{d i_{L}}{d t} \\
\frac{d i_{L}}{d t}=\frac{V_{s}-V_{o}}{L}
\end{array}
$$

The rate of change of inductor current is a constant, so the current must change linearly while the switch is open. The change in inductor current while the switch is open is

$$
\frac{\Delta i_{L}}{\Delta t}=\frac{\Delta i_{L}}{(1-D) T}=\frac{V_{s}-V_{o}}{L}
$$

Solving for $\Delta i_{L}$

$$
\left(\Delta i_{L}\right)_{\text {open }}=\frac{\left(V_{s}-V_{o}\right)(1-D) T}{L}
$$ [Eq. 06026]

For steady-state operation, the net change in inductor current must be zero. Using Eqs. 06025 and 06026,

$$
\begin{array}{c}
\left(\Delta i_{L}\right)_{\text {closed }}+\left(\Delta i_{L}\right)_{\text {open }}=0 \\
\frac{V_{s} D T}{L}+\frac{\left(V_{s}-V_{o}\right)(1-D) T}{L}=0
\end{array}
$$

Solving for $V_{o}$,

$$
V_{s}(D+1-D)-V_{o}(1-D)=0
$$

$$
\boxed{V_{o}=\frac{V_{s}}{1-D}}
$$ [Eq. 06027]

Also, the average inductor voltage must be zero for periodic operation. Expressing the average inductor voltage over one switching period,

$$
V_{L}=V_{s} D+\left(V_{s}-V_{o}\right)(1-D)=0
$$

Solving for $V_{o}$ yields the same result as in Eq. 06027 Equation 06027 shows that if the switch is always open and $D$ is zero, the output voltage is the same as the input. As the duty ratio is increased, the denominator of Eq. 06027 becomes smaller, resulting in a larger output voltage. The boost converter produces an output voltage that is greater than or equal to the input voltage. However, the output voltage cannot be less than the input, as was the case with the buck converter.

As the duty ratio of the switch approaches $1$, the output voltage goes to infinity according to Eq. 06027. However, Eq. 06027 is based on ideal components. Real components that have losses will prevent such an occurrence, as shown later in this section. Figure 06009 shows the voltage and current waveforms for the boost converter.

The average current in the inductor is determined by recognizing that the average power supplied by the source must be the same as the average power absorbed by the load resistor. Output power is

$$
P_{o}=\frac{V_{o}^{2}}{R}=V_{o} I_{o}
$$

and input power is $V_{s} I_{s}=V_{s} I_{L}$. Equating input and output powers and using Eq. 06027

$$
V_{S} I_{L}=\frac{V_{o}^{2}}{R}=\frac{\left[V_{S} /(1-D)\right]^{2}}{R}=\frac{V_{s}^{2}}{(1-D)^{2}} R
$$

By solving for average inductor current and making various substitutions, $I_{L}$ can be expressed as

$$
\boxed{I_{L}=\frac{V_{s}}{(1-D)^{2} R}=\frac{V_{o}^{2}}{V_{s} R}=\frac{V_{o} I_{o}}{V_{s}}}
$$ [Eq. 06028]

Maximum and minimum inductor currents are determined by using the average value and the change in current from Eq. 06025.

$$
\begin{aligned}
I_{\max } &=I_{L}+\frac{\Delta i_{L}}{2}=\frac{V_{s}}{(1-D)^{2} R}+\frac{V_{s} D T}{2 L} \\
\end{aligned}
$$ [Eq. 06029]

$$
\begin{aligned}
I_{\min } &=I_{L}-\frac{\Delta i_{L}}{2}=\frac{V_{s}}{(1-D)^{2} R}-\frac{V_{s} D T}{2 L} \\
\end{aligned}
$$ [Eq. 06030]

Equation 06027 was developed with the assumption that the inductor current is continuous, meaning that it is always positive. A condition necessary for continuous inductor current is for $I_{\min }$ to be positive. Therefore, the boundary between continuous and discontinuous inductor current is determined from


$$
\begin{aligned}
I_{\min }=0=\frac{V_{s}}{(1-D)^{2} R}-\frac{V_{s} D T}{2 L} \\
\end{aligned}
$$

or

$$
\begin{aligned}
\frac{V_{s}}{(1-D)^{2} R}=\frac{V_{s} D T}{2 L}=\frac{V_{s} D}{2 L f} \\
\end{aligned}
$$

The minimum combination of inductance and switching frequency for continuous current in the boost converter is therefore

$$
(L f)_{\min }=\frac{D(1-D)^{2} R}{2}
$$ [Eq. 06031]

or

$$
\boxed{L_{\min }=\frac{D(1-D)^{2} R}{2 f}}
$$ [Eq. 06032]

A boost converter designed for __continuous-current operation__ will have an inductor value greater than $L_{\min }$.

From a design perspective, it is useful to express $L$ in terms of a desired $\Delta i_{L}$

$$
L=\frac{V_{s} D T}{\Delta i_{L}}=\frac{V_{s} D}{\Delta i_{L} f}
$$ [Eq. 06033]

## Output Voltage Ripple

The preceding equations were developed on the assumption that the output voltage was a constant, implying an infinite capacitance. In practice, a finite capacitance will result in some fluctuation in output voltage, or ripple.

The peak-to-peak output voltage ripple can be calculated from the capacitor current waveform, shown in Fig. 06009d. The change in capacitor charge can be calculated from

$$
|\Delta Q|=\left(\frac{V_{o}}{R}\right) D T=C \Delta V_{o}
$$

An expression for ripple voltage is then

$$
\begin{aligned}
\Delta V_{o}=\frac{V_{o} D T}{R C}=\frac{V_{o} D}{R C f} \\
\end{aligned}
$$

or

$$
\begin{aligned}
\boxed{\frac{\Delta V_{o}}{V_{o}}=\frac{D}{R C f}} \\
\end{aligned}
$$ [Eq. 06034]

where $f$ is the switching frequency. Alternatively, expressing capacitance in terms of output voltage ripple yields

$$
C=\frac{D}{R\left(\Delta V_{o} / V_{o}\right) f}
$$ [Eq. 06035]

As with the buck converter, equivalent series resistance of the capacitor can contribute significantly to the output voltage ripple. The peak-to-peak variation in capacitor current Fig. 06009 is the same as the maximum current in the inductor. The voltage ripple due to the ESR is

$$
\Delta V_{o, \mathrm{ESR}}=\Delta i_{C^{r} C}=I_{L, \max } r_{C}
$$ [Eq. 06036]

## Inductor Resistance

Inductors should be designed to have small resistance to minimize power loss and maximize efficiency. The existence of a small inductor resistance does not substantially change the analysis of other converters like the buck as example. However, inductor resistance princiopally affects performance of the boost converter, and especially at high duty ratios.

For the boost converter, the transfer function is:

$$
\frac{V_{o}}{V_{s}}=\frac{1}{1-D}
$$ [Eq. 06037}

To understand the effect of inductor resistance on the output voltage, assume that the inductor current is approximately constant. The source current is the same as the inductor current, and average diode current is the same as average load current. The power supplied by the source must be the same as the power absorbed by the load and the inductor resistance, neglecting other losses.

$$
P_{s}=P_{o}+P_{r_{L}}
$$ [Eq. 06038]

$$
V_{s} I_{L}=V_{o} I_{D}+I_{L}^{2} r_{L}
$$

where $r_{L}$ is the series resistance of the inductor. The diode current is equal to the inductor current when the switch is off and is zero when the switch is on. Therefore, the average diode current is given by: 

$$
I_{D}=I_{L}(1-D)
$$  [Eq. 06039]

Substituting for $I_{D}$ into last Eq. 06038 and solving for $V_{s}$ gives:

$$
V_{s} I_{L}=V_{o} I_{L}(1-D)+I_{L}^{2} r_{L}
$$

which becomes

$$
V_{s}=V_{o}(1-D)+I_{L} r_{L}
$$ [Eq. 06040]

Now considering $I_{L}$ from Eq. 06039, in terms of Vo:

$$
I_{L}=\frac{I_{D}}{1-D}=\frac{V_{o} / R}{1-D}
$$ [Eq. 06041]

Substituting for $I_{L}$ into Eq. [06040]:

$$
V_{s}=\frac{V_{o} r_{L}}{R(1-D)}+V_{o}(1-D)
$$

From the las equation, Solving for $V_{o}$ the Transfer function is given by:

$$
\frac{V_{o}}{V_{s}}=\left(\frac{1}{1-D}\right)\left(\frac{1}{1+r_{L} /\left[R(1-D)^{2}\right]}\right)
$$ [Eq. 06042]

The preceding equation is similar to that for an ideal converter but includes a correction factor to account for the inductor resistance. The next Figure 06010a shows the output voltage of the boost converter with and without inductor resistance.

The inductor resistance also has an effect on the power efficiency of converters. Efficiency is the ratio of output power to output power plus losses. For the boost converter:

$$
\eta=\frac{P_{o}}{P_{o}+P_{\text { loss }}}
$$ [Eq. 06043]

Using Eq. 06041 for $I_{L}$,

$$
\eta=\frac{V_{o}^{2} / R}{V_{o}^{2} / R+\left(V_{o} / R\right)^{2} /(1-D) r_{L}}=\frac{1}{1+r_{L}\left[R(1-D)^{2}\right]}
$$ [Eq. 06044]

As the duty ratio increases, the efficiency of the boost converter decreases, as indicated in Fig. 06010b.

## Boost Converter with Discontinuous Current


The boost converter will also operate for discontinuous inductor current. In some cases, the discontinuous-current mode is desirable for control reasons in the case of a regulated output. The relationship between output and input voltages is determined from two relationships:

   1. The average inductor voltage is zero.
   2. The average current in the diode is the same as the load current.

The inductor and diode currents for discontinuous current have the basic waveforms as shown in Fig. 06022a and 06022c. When the switch is on, the voltage across the inductor is $V_{s}$. When the switch is off and the inductor current is positive, the inductor voltage is $V_{s}-V_{o}$. The inductor current decreases until it reaches zero and is prevented from going negative by the diode. With the switch open and the diode off, the inductor current is zero. The average voltage across the inductor is

$$
V_{s} D T+\left(V_{s}-V_{o}\right) D_{1} T=0
$$

which results in

$$
V_{o}=V_{s}\left(\frac{D+D_{1}}{D_{1}}\right)
$$ [Eq. 06097]

The average diode current Fig. 06022c is

$$
I_{D}=\frac{1}{T}\left(\frac{1}{2} I_{\max } D_{1} T\right)=\frac{1}{2} I_{\max } D_{1}
$$ [Eq. 06098]

<p align="center"><img src="images/Discontinuous_Inductor_Current_Boost_Converter.png" style="width:70%"></p>
<center> Figure 06022. Discontinuous current in the boost converter.
(a) Inductor current; (b) Inductor voltage; (c) Diode current. </center>
<br><br/>

Current $I_{\max }$ is the same as the change in inductor current when the switch is closed.

$$
I_{\max }=\Delta i_{L}=\frac{V_{S} D T}{L}
$$ [Eq. 06099]

Substituting for $I_{\max }$ in Eq. 06098 and setting the result equal to the load current,

$$
I_{D}=\frac{1}{2}\left(\frac{V_{s} D T}{L}\right) D_{1}=\frac{V_{o}}{R}
$$ [Eq. 06100]

Solving for $D_{1}$

$$
D_{1}=\left(\frac{V_{o}}{V_{s}}\right)\left(\frac{2 L}{R D T}\right)
$$ [Eq. 06101]

Substituting the preceding expression for $D_{1}$ into Eq. 06097 results in the quadratic equation

$$
\left(\frac{V_{o}}{V_{s}}\right)^{2}-\frac{V_{o}}{V_{s}}-\frac{D^{2} R T}{2 L}=0
$$

Solving for $V_{\mathrm{o}} / V_{\mathrm{s}}$

$$
\frac{V_{o}}{V_{s}}=\frac{1}{2}\left(1+\sqrt{1+\frac{2 D^{2} R T}{L}}\right)
$$ [Eq. 06102]

The boundary between continuous and discontinuous current occurs when $D_{1}= 1-D$. Another condition at the boundary is when $I_{\min }$ in Eq. 06030 is zero.

Whether the boost converter is operating in the continuous or discontinuous mode depends on the combination of circuit parameters, including the duty ratio. As the duty ratio for a given boost converter is varied, the converter may go into and out of the discontinuous mode.
