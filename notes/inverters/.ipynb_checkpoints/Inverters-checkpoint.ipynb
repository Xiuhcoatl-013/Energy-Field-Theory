# Inverters
## Converting DC to AC


## Introduction
Inverters are circuits that convert dc to ac. More precisely, inverters transfer power from a dc source to an ac load. From applications point of view, the objective is to create an ac voltage when only a dc voltage source is available. Inverters are used in applications such as adjustable-speed ac motor drives, uninterruptible power supplies (UPS), and running ac appliances from an battery.

## The Full-Bridge Converter
The full-bridge converter of Fig 08001a is the basic circuit used to convert dc to ac. The full-bridge converter was introduced as part of a dc power supply circuit. In this application, an ac output is synthesized from a dc input by closing and opening the switches in an appropriate sequence. The output voltage $v_{o}$ can be $+V_{\mathrm{dc}}$, $-V_{\mathrm{dc}}$, or zero, depending on which switches are closed. Figure 08001b to 1e shows the equivalent circuits for switch combinations.

<br><br/>
$$
\begin{array}{ll}
\hline \text { Switches Closed } & \text { Output Voltage } v_{o} \\
\hline S_{1} \text { and } S_{2} & +V_{\mathrm{dc}} \\
S_{3} \text { and } S_{4} & -V_{\mathrm{dc}} \\
S_{1} \text { and } S_{3} & 0 \\
S_{2} \text { and } S_{4} & 0 \\
\hline
\end{array}
$$
<br><br/>

<br><br/>
<p align="center"><img src="images/Full-bridge_Inverter.png" style="width:50%"></p>
<center> Figure 08001.$\quad$ (a) Full-bridge Inverter; (b) $S_{1}$ and $S_{2}$ closed; (c) $S_{3}$ and $S_{4}$ closed; (d) $S_{1}$ and $S_{3}$ closed; (e) $S_{2}$ and $S_{4}$ closed. </center>
<br><br/>
<br><br/>

Note that S1 and S4 should not be closed at the same time, nor should S2 and S3. Otherwise, a short circuit would exist across the dc source. Real switches do not turn on or off instantaneously, as was discussed before. Therefore, switching transition times must be accommodated in the control of the switches. Overlap of switch “on” times will result in a short circuit, sometimes called a shoot-through fault, across the dc voltage source. The time allowed for switching
is called blanking time.


## The Square-Wave Inverter

The simplest switching scheme for the full-bridge converter produces a square wave output voltage. The switches connect the load to $+V_{\mathrm{dc}}$ when $S_{1}$ and $S_{2}$ are closed or to $-V_{\mathrm{dc}}$ when $S_{3}$ and $S_{4}$ are closed. The periodic switching of the load voltage between $+V_{\mathrm{dc}}$ and $-V_{\mathrm{dc}}$ produces a square wave voltage across the load. Although this alternating output is nonsinusoidal, it may be an adequate ac waveform for some applications.
The current waveform in the load depends on the load components. For the resistive load, the current waveform matches the shape of the output voltage. An inductive load will have a current that has more of a sinusoidal quality than the voltage because of the filtering property of the inductance. An inductive load presents some considerations in designing the switches in the full-bridge circuit because the switch currents must be bidirectional.
For a series RL load and a square wave output voltage, assume switches $S_{1}$ and $S_{2}$ in Fig. 08001a close at $t = 0$. The voltage across the load is $+V_{\mathrm{dc}}$, and current begins to increase in the load and in $S_{1}$ and $S_{2}$. The current is expressed as the sum of the forced and natural responses.

$$
\begin{aligned} i_{o}(t) &=i_{f}(t)+i_{n}(t) \\ &=\frac{V_{\mathrm{dc}}}{R}+A e^{-t / \tau} \quad \text { for } 0 \leq t \leq T / 2 \end{aligned}
$$   [Eq. 08001]

where $A$ is a constant evaluated from the initial condition and $\tau=L / R$. 

At $t=T / 2$, $S_{1}$ and $S_{2}$ open, and $S_{3}$ and $S_{4}$ close. The voltage across the RL load becomes $-V_{\mathrm{dc}}$, and the current has the form

$$
i_{o}(t)=\frac{-V_{\mathrm{dc}}}{R}+B e^{-(t-T / 2) / \tau} \quad \text { for } \quad T / 2 \leq t \leq T
$$   [Eq. 08002]

where the constant $B$ is evaluated from the initial condition.
When the circuit is first energized and the initial inductor current is zero, a transient occurs before the load current reaches a steady-state condition. At steady state, $i$ is periodic and symmetric about zero, as illustrated in Fig. 08002. Let the initial condition for the current described in [Eq. 08001] be $I_{\min }$, and let the initial condition for the current described in [Eq. 08002] be $I_{\max }$. Evaluating [Eq. 08001] at $t=0$,

$$
i_{o}(0)=\frac{V_{\mathrm{dc}}}{R}+A e^{0}=I_{\mathrm{min}}
$$

or

$$
A=I_{\min }-\frac{V_{\mathrm{dc}}}{R}
$$   [Eq. 08003]

<br><br/>
<p align="center"><img src="images/Square_Wave_Output_Voltage.png" style="width:50%"></p>
<center> Figure 08002.$\quad$ Square wave output voltage and steady-state current
waveform for an RL load. </center>
<br><br/>
<br><br/>

Likewise, [Eq. 08002] is evaluated at $t=T / 2$

$$
i_{o}(T / 2)=\frac{-V_{\mathrm{dc}}}{R}+B e^{0}=I_{\max }
$$

or

$$
B=I_{\max }+\frac{V_{\mathrm{dc}}}{R}
$$   [Eq. 08004]

In steady state, the current waveforms described by [Eq. 08001] and [Eq. 08002] then become

$$
i_{o}(t)=\left\{\begin{array}{ll}\frac{V_{\mathrm{dc}}}{R}+\left(I_{\min }-\frac{V_{\mathrm{dc}}}{R}\right) e^{-t / \tau} & \text { for } 0<t<\frac{T}{2} \\ \frac{-V_{\mathrm{dc}}}{R}+\left(I_{\max }+\frac{V_{\mathrm{dc}}}{R}\right)e^{(t-T / 2) / \tau} & \text { for } \frac{T}{2}<t<T\end{array}\right.
$$   [Eq. 08005]

An expression is obtained for $I_{\max }$ by evaluating the first part of [Eq. 08005] at $t=T / 2$

$$
i(T / 2)=I_{\max }=\frac{V_{\mathrm{dc}}}{R}+\left(I_{\min }-\frac{V_{\mathrm{dc}}}{R}\right) e^{-(T / 2 \tau)}
$$   [Eq. 08006]   

and by symmetry,

$$
I_{\min }=-I_{\max }
$$   [Eq. 08007]

Substituting $-I_{\max }$ for $-I_{\min }$ in [Eq. 08006] and solving for $I_{\max }$,

$$
I_{\max }=-I_{\min }=\frac{V_{\mathrm{dc}}}{R}\left(\frac{1-e^{-T / 2 \tau}}{1+e^{-T / 2 \tau}}\right)
$$   [Eq. 08008]

Thus, [Eq. 08005] and [Eq. 08008] describe the current in an RL load in the steady state when a square wave voltage is applied. Figure 08002 shows the resulting currents in the load, source, and switches.
Power absorbed by the load can be determined from $I_{\mathrm{rms}}^{2} R$, where rms load current is determined from the defining equation. The integration may be simplified by taking advantage of the symmetry of the waveform. Since the square each of the current half-periods is identical, only the first half-period
needs to be evaluated:

$$
I_{\mathrm{rms}}=\sqrt{\frac{1}{T} \int_{0}^{T} i^{2}(t) d(t)}=\sqrt{\frac{2}{T} \int_{0}^{T / 2}\left[\frac{V_{\mathrm{dc}}}{R}+\left(I_{\min }-\frac{V_{\mathrm{dc}}}{R}\right) e^{-t / \tau}\right]^{2} d t}
$$   [Eq. 08009]

If the switches are ideal, the power supplied by the source must be the same as absorbed by the load. Power from a dc source is determined from

$$
P_{\mathrm{dc}}=V_{\mathrm{dc}} I_{s}
$$   [Eq. 08010]

The switch currents in Fig. 08002 show that the switches in the full-bridge circuit must be capable of carrying both positive and negative currents for $RL$ loads. However, real electronic devices may conduct current in one direction only. This problem is solved by placing feedback diodes in parallel (anitparallel) with each switch. During the time interval when the current in the switch must be negative, the feedback diode carries the current. The diodes are reverse-biased when current is positive in the switch. Figure 08003a shows the full-bridge inverter with switches implemented as insulated gate bipolar transistors (IGBTs) with feedback diodes. Transistor and diode currents for a square wave voltage and an $RL$ load are indicated in Fig 08003b. Power semiconductor modules often include feedback diodes with the switches.

<br><br/>
<p align="center"><img src="images/Full-bridge_Inverter_Using_IGBTs.png" style="width:50%"></p>
<center> Figure 08003.$\quad$ (a) Full-bridge inverter using IGBTs;
(b) Steady-state current for an $RL$ load. </center>
<br><br/>
<br><br/>

When IGBTs $Q_{1}$ and $Q_{2}$ are turned off in Fig. 08003a, the load current must be continuous and will transfer to diodes $D_{3}$ and $D_{4}$, making the output voltage $-V_{\mathrm{dc}}$, effectively turning on the switch paths 3 and 4 before $Q_{3}$ and $Q_{4}$ are turned on. IGBTs $Q_{3}$ and $Q_{4}$ must be turned on before to permit the load current decays to zero.

## Fourier Series Analysis

The Fourier series method is often the most practical way to analyze load current and to compute power absorbed in a load, especially when the load is more complex than a simple resistive or RL load. Auseful approach for inverter analysis is to express the output voltage and load current in terms of a Fourier series. __With no dc component in the output,__

$$
v_{o}(t)=\sum_{n=1}^{\infty} V_{n} \sin \left(n \omega_{0} t+\theta_{n}\right)
$$   [Eq. 08011]

and

$$
i_{o}(t)=\sum_{n=1}^{\infty} I_{n} \sin \left(n \omega_{0} t+\phi_{n}\right)
$$   [Eq. 08012]

Power absorbed by a load with a series resistance is determined from $I_{\mathrm{ms}}^{2} R$, where the rms current can be determined from the rms currents at each of the components in the Fourier series by

$$
I_{\mathrm{rms}}=\sqrt{\sum_{n=1}^{\infty} I_{n, \mathrm{rms}}^{2}}=\sqrt{\sum_{n=1}^{\infty}\left(\frac{I_{n}}{\sqrt{2}}\right)^{2}}
$$   [Eq. 08013]

where

$$
I_{n}=\frac{V_{n}}{Z_{n}}
$$   [Eq. 08014]

and $Z_{n}$ is the load impedance at harmonic n.
Equivalently, the power absorbed in the load resistor can be determined for
each frequency in the Fourier series. Total power can be determined from

$$
P=\sum_{n=1}^{\infty} P_{n}=\sum_{n=1}^{\infty} I_{n, \mathrm{ms}}^{2} R
$$   [Eq. 08015]

where $I_{n, \mathrm{mns}}$ is $I_{n} / \sqrt{2}$

In the case of the square wave, the Fourier series contains the odd harmonics
and can be represented as

$$
v_{o}(t)=\sum_{n \text { odd }} \frac{4 V_{\mathrm{dc}}}{n \pi} \sin n \omega_{0} t
$$   [Eq. 08016]

## Total Harmonic Distortion

Since the objective of the inverter is to use a dc voltage source to supply a load requiring ac, it is useful to describe the quality of the ac output voltage generated.
The quality of a nonsinusoidal wave can be expressed in terms of total harmonic distortion (THD), defined previously. Assuming no dc component in the output,

$$
\mathrm{THD}=\frac{\sqrt{\sum_{n=2}^{\infty}\left(V_{n, \mathrm{ms}}\right)^{2}}}{V_{1, \mathrm{rms}}}=\frac{\sqrt{V_{\mathrm{rms}}^{2}-V_{1, \mathrm{rms}}^{2}}}{V_{1, \mathrm{rms}}}
$$   [Eq. 08017]


The THD of current is determined by substituting current for voltage in the above equation. The THD of load current is often of greater interest than that of output voltage. __Becouse the definition for THD is based on the Fourier series__, then by obvious, there is some benefit in using the Fourier series method for analysis when the THD must be determined. Other measures of distortion such as distortion factor, as presented previously, can also be applied to describe the output waveform for inverters. ___[Take in consideration that the quality of the power signal play an important role in the power factor (PF) and then in the overal performance of the system from an energy eficiency point of view]___

## Amplitude and Harmonic Control

The amplitude of the fundamental frequency for a square wave output from of
the full-bridge inverter is determined by the dc input voltage [Eq. 08016]. A controlled output can be produced by modifying the switching scheme. An output voltage of the form shown in Fig. 08005a has intervals when the output is zero as well as $+V_{\mathrm{dc}}$ and $-V_{\mathrm{dc}}$. This output voltage can be controlled by adjusting the interval $\alpha$ on each side of the pulse where the output is zero. The rms value of the voltage waveform in Fig 08005a is

$$
V_{\mathrm{rms}}=\sqrt{\frac{1}{\pi} \int_{\alpha}^{\pi - \alpha} V_{\mathrm{dc}}^{2} d(\omega t)}=V_{\mathrm{dc}} \sqrt{1-\frac{2 \alpha}{\pi}}
$$   [Eq. 08018]

<br><br/>
<p align="center"><img src="images/Inverter_Output_for_Amplitude_and_Harmonic_Control.png" style="width:50%"></p>
<center> Figure 08005.$\quad$ a) Inverter output for amplitude and harmonic control; (b) Switching sequence for the full-bridge inverter of Fig. 1a. </center>
<br><br/>
<br><br/>

The Fourier series of the waveform is expressed as

$$
v_{o}(t)=\sum_{n \text { odd }} V_{n} \sin \left(n \omega_{0} t\right)
$$   [Eq. 08019]

Taking advantage of half-wave symmetry, the amplitudes are

$$
V_{n}=\frac{2}{\pi} \int_{\alpha}^{\pi-\alpha} V_{\mathrm{dc}} \sin \left(n \omega_{0} t\right) d\left(\omega_{0} t\right)=\frac{4 V_{\mathrm{dc}}}{n \pi} \cos (n \alpha)
$$   [Eq. 08020]

where $\alpha$ is the angle of zero voltage on each end of the pulse. The amplitude of each frequency of the output is a function of $\alpha$. In particular, the amplitude of the fundamental frequency $(n=1)$ is controllable by adjusting $\alpha$:

$$
V_{1}=\left(\frac{4 V_{\mathrm{dc}}}{\pi}\right) \cos \alpha
$$   [Eq. 08021]

Harmonic content can also be controlled by adjusting $\alpha$. If $\alpha=30^{\circ}$, for example, $V_3 = 0$. This is significant because the third harmonic can be eliminated from the output voltage and current. Other harmonics can be eliminated by choosing a value of $\alpha$ which makes the cosine term in [Eq. 08020] to go to zero.
Harmonic $n$ is eliminated if

$$
\alpha=\frac{90^{\circ}}{n}
$$   [Eq. 08022]

The switching scheme required to produce an output like Fig. 08005a must provide intervals when the output voltage is zero, as well as $\pm V_{\mathrm{dc}}$. The switching sequence of Fig. 08005b is a way to implement the required output waveform. 

Amplitude control and harmonic reduction may not be compatible. For
example, establishing $\alpha$ at $30^{\circ}$ to eliminate the third harmonic fixes the amplitude of the output fundamental frequency at $V_{1}=\left(4 V_{\mathrm{dc}} / \pi\right) \cos 30^{\circ}=1.1 V_{\mathrm{dc}}$ and
removes further controllability. To control both amplitude and harmonics using this switching scheme, it is necessary to be able to control the dc input voltage to the inverter. A dc-dc converter placed between the dc source and the inverter can provide a controlled dc input to the inverter.

A graphical representation of the integration in the Fourier series coefficient of [Eq. 08020] gives some insight into harmonic elimination. Recall from previous theory that the Fourier coefficients are determined from the integral of the product of the waveform and a sinusoid. Figure 08006a shows the output waveform for $\alpha=30^{\circ}$ and the sinusoid of $\omega=3 \omega_{o}$. The product of these two waveforms has an area of zero, showing that the third harmonic is zero. Figure 6b shows the waveform for $\alpha=18^{\circ}$ and the sinusoid of $\omega=5 \omega_{o}$, showing that the fifth harmonic is eliminated for this value of $\alpha$.
Other switching schemes can eliminate multiple harmonics. For example, the output waveform shown in Fig. 08006c eliminates both the third and fifth harmonics, as indicated by the areas of both being zero.

## MULTILEVEL INVERTERS

The $\mathrm{H}$ bridge inverter previously illustrated in Figs. 08001 and 08003 produces output voltages of $V_{\mathrm{dc}}, 0,$ and $-V_{\mathrm{dc}} .$ The basic H bridge switching concept can be expanded to other circuits that can produce additional output voltage levels.

<br><br/>
<p align="center"><img src="images/inverter_with_two_dc_sources.png" style="width:50%"></p>
<center> Figure 08009.$\quad$ An inverter with two dc sources, each with an
H bridge implemented with IGBTs. </center>
<br><br/>
<br><br/>

These multilevel-output voltages are more sinelike in quality and thus reduce harmonic content. The multilevel inverter is suitable for applications including adjustable-speed motor drives and interfacing renewable energy sources such as photovoltaics to the electric power grid.

__Multilevel Converters with Independent DC Sources__
One multilevel inverter method uses independent dc sources, each with an H bridge. A circuit with two dc voltage sources is shown in Fig. 08009. The output of each of the $\mathrm{H}$ bridges is $+V_{\mathrm{dc}}$, $-V_{\mathrm{d} \mathrm{s}}$, or $0$, as was illustrated in Fig. 08001. The total instantaneous voltage $v_{o}$ on the output of the multilevel converter is any combination of individual bridge voltages. Thus, for a two-source inverter, $v_{o}$ can be any of the five levels $+2 V_{\mathrm{dc}}, V_{\mathrm{dc}}, 0,-V_{\mathrm{dc}}$, or $-2 V_{\mathrm{dc}}$
Each H bridge operates with a switching scheme like that of Fig. 08005 which was used for amplitude or harmonic control. Each bridge operates at a different delay angle $\alpha$, resulting in bridge and total output voltages like those shown in Fig. 08010.

The Fourier series for the total output voltage $v_{o}$ for the two-source circuit contains only the odd-numbered harmonics and is

$$
v_{o}(t)=\frac{4 V_{\mathrm{dc}}}{\pi} \sum_{n=1,3,5,7, \ldots}^{\infty}\left[\cos \left(n \alpha_{1}\right)+\cos \left(n \alpha_{2}\right)\right] \frac{\sin \left(n \omega_{0} t\right)}{n}
$$   [Eq. 08023]

<br><br/>
<p align="center"><img src="images/h_bridges_voltage_output.png" style="width:50%"></p>
<center> Figure 08010.$\quad$ Voltage output of each of the H bridges and
the total voltage for the two-source multilevel inverter of Fig. 08009. </center>
<br><br/>
<br><br/>

The Fourier coefficients for this series are

$$
V_{n}=\frac{4 V_{\mathrm{dc}}}{n \pi}\left[\cos \left(n \alpha_{1}\right)+\cos \left(n \alpha_{2}\right)\right]
$$   [Eq. 08024]

The modulation index $M_{i}$ is the ratio of the amplitude of the fundamental frequency component of $v_{o}$ to the amplitude of the fundamental frequency component of a square-wave of amplitude $2 V_{\mathrm{dc}},$ which is $2\left(4 V_{\mathrm{dc}} / \pi\right)$.

$$
M_{t}=\frac{V_{1}}{2\left(4 V_{\mathrm{dc}} / \pi\right)}=\frac{\cos \alpha_{1}+\cos \alpha_{2}}{2}
$$   [Eq. 08025]

Some harmonics can be eliminated from the output voltage waveform with the proper selection of $\alpha_{1}$ and $\alpha_{2}$ in [Eq. 08024]. For the two-source converter, harmonic $m$ can be eliminated by using delay angles such that

$$
\cos \left(m \alpha_{1}\right)+\cos \left(m \alpha_{2}\right)=0
$$   [Eq. 08026]

To eliminate the $m$ th harmonic and also meet a specified modulation index for the two-source inverter requires the simultaneous solution to [Eq. 08026] and the additional equation derived from [Eq. 08025],

$$
\cos \left(\alpha_{1}\right)+\cos \left(\alpha_{2}\right)=2 M_{t}
$$   [Eq. 08027]

To solve [Eq. 08026] and [Eq. 08027] simultaneously requires an iterative numerical method such as the Newton-Raphson method.

The preceding concept can be extended to a multilevel converter having several de sources. For $k$ separate sources connected in cascade, there are $2 k+1$ possible voltage levels. As more de sources and $\mathrm{H}$ bridges are added, the total output voltage has more steps, producing a staircase waveform that more closely approaches a sinusoid. For a five-source system as shown in Fig. 08011, there are 11 possible output voltage levels, as illustrated in Fig. 08012

The Fourier series for a staircase waveform such as that in Fig. 08012 for $k$ separate de sources each equal to $V_{\mathrm{dc}}$ is

$$
v_{o}(t)=\frac{4 V_{\mathrm{dc}}}{\pi} \sum_{n=1}^{\infty} \sum_{3,5,7, \ldots}\left[\cos \left(n \alpha_{1}\right)+\cos \left(n \alpha_{2}\right)+\cdots+\cos \left(n \alpha_{k}\right)\right] \frac{\sin \left(n \omega_{0} t\right)}{n}(8-28)
$$   [Eq. 08028]

The magnitudes of the Fourier coefficients are thus

$$
\begin{array}{c}
V_{n}=\frac{4 V_{\mathrm{dc}}}{n \pi}\left[\cos \left(n \alpha_{1}\right)+\cos \left(n \alpha_{2}\right)+\cdots+\cos \left(n \alpha_{k}\right)\right] \\
\text { for } n=1,3,5,7, \ldots
\end{array}
$$   [Eq. 08029]

<br><br/>
<p align="center"><img src="images/five-source_cascade_multilevel_converter.png" style="width:50%"></p>
<center> Figure 08011.$\quad$ A five-source cascade multilevel converter. </center>
<br><br/>
<br><br/>

The modulation index $M_{i}$ for $k$ dc sources each equal to $V_{\mathrm{dc}}$ is

$$
M_{I}=\frac{V_{1}}{4 k V_{\mathrm{d} \mathrm{d}} / \pi}=\frac{\cos \left(\alpha_{1}\right)+\cos \left(\alpha_{2}\right)+\cdots+\cos \left(\alpha_{k}\right)}{k}
$$   [Eq. 08030]

$$
\cos \left(m \alpha_{1}\right)+\cos \left(m \alpha_{2}\right)+\cdots+\cos \left(m \alpha_{k}\right)=0
$$   [Eq. 08031]

__Equalizing Average Source Power with Pattern Swapping__
In the two-source inverter of Fig. 08009 using the switching scheme of Fig. 08010, the source and $\mathrm{H}$ bridge producing the voltage $v_{1}$ supplies more average power (and energy) than the source and $\mathrm{H}$ bridge producing $v_{2}$ due to longer pulse widths in both the positive and negative half cycles. If the dc sources are batteries, one battery will discharge faster than the other. A technique known as pattern swapping or duty swapping equalizes the average power supplied by each dc source.

The principle of pattern swapping is to have each dc source conduct for an equal amount of time on average. An alternate switching scheme for the twosource circuit is shown in Fig. 08013. In this scheme, the first source conducts for a longer time in the first half-cycle while the second source conducts for more time in the second half-cycle. Thus, over one complete period, the sources conduct equally, and average power from each source is the same.

For the five-source converter in Fig. 08011, a switching scheme to equalize average power is shown in Fig. 08014. Note that five half cycles are required to equalize power.

A variation of the $\mathrm{H}$ bridge multilevel inverter is to have the dc sources be of different values. The output voltage would be a staircase waveform, but not in equal voltage increments. The Fourier series of the output voltage would have differentvalued harmonic amplitudes which may be an advantage in some applications. Because independent voltage sources are needed, the multiple-source implementation of multilevel converters is best suited in applications where batteries, fuel cells, or photovoltaics are the sources.

__Diode-Clamped Multilevel Inverters__
A multilevel converter circuit that has the advantage of using a single dc source rather than multiple sources is the diode-clamped multilevel converter shown in Fig. 08015a. In this circuit, the de voltage source is connected to a pair of series capacitors, each charged to $V_{\mathrm{de}} / 2 .$ The following analysis shows how the output voltage can have the levels of $V_{\mathrm{dc}}, V_{\mathrm{dc}} / 2,0,-V_{\mathrm{de}} / 2,$ or $-V_{\mathrm{dc}}$

<br><br/>
<p align="center"><img src="images/diode-clamped_multilevel_inverter.png" style="width:50%"></p>
<center> Figure 08015.$\quad$ (a) A diode-clamped multilevel inverter implemented with IGBTs. (b) Analysis for one-half of the circuit for $v_{1}=V_{\mathrm{d} c},(c)$ for $v_{1}=0,$ and $(d)$ for $v_{1}=\frac{1}{2} V_{\mathrm{d} c}$ </center>
<br><br/>
<br><br/>

For the analysis, consider only the left half of the bridge, as shown in Fig. 08015b, c, and d. With $S_{1}$ and $S_{2}$ closed and $S_{3}$ and $S_{4}$ open, $V_{1}=V_{\mathrm{dc}}$ (Fig. 08015b). The diodes are off for this condition. With $S_{1}$ and $S_{2}$ open and $S_{3}$ and $S_{4}$ closed, $V_{1}=0$ (Fig. 08015c). The diodes are off for this condition also. To produce a voltage of $V_{\mathrm{dc}} / 2, S_{2}$ and $S_{3}$ are closed, and $S_{1}$ and $S_{4}$ are open (Fig. 08015d). The voltage $v_{1}$ is that of the lower capacitor, at voltage $V_{\mathrm{dc}} / 2,$ connected through the antiparallel diode path that can carry load current in either direction. Note that for each of these circuits, two switches are open, and the voltage of the source divides between the two, thus reducing the voltage stress across each switch compared to the H bridge circuit of Fig. 08001

Using a similar analysis, the right half of the bridge can also produce the voltages $V_{\mathrm{dc}}, 0,$ and $V_{\mathrm{de}} / 2$. The output voltage is the difference of the voltages between each half bridge, resulting in the five levels

$$
v_{o} \in\left\{V_{\mathrm{dc}}, \frac{1}{2} V_{\mathrm{dc}}, 0,-\frac{1}{2} V_{\mathrm{dc}},-V_{\mathrm{dc}}\right\}
$$   [Eq. 08032]

with multiple ways to achieve some of them. The switch control can establish delay angles $\alpha_{1}$ and $\alpha_{2}$, to produce an output voltage like that in Fig. 08010 for the cascaded H bridge, except that the maximum value is $V_{\mathrm{dc}}$ instead of $2 V_{\mathrm{dc}^{*}}$

More output voltage levels are achieved with additional capacitors and switches. Figure 08016 shows the de source divided across three series capacitors.

<br><br/>
<p align="center"><img src="images/clamped_multilevel_inverter_4x_v_levels.png" style="width:50%"></p>
<center> Figure 08016.$\quad$ A diode-clamped multilevel inverter that produces four voltage levels on each side of the bridge and seven output
voltage levels. </center>
<br><br/>
<br><br/>

The voltage across each capacitor is $\frac{1}{3} \mathrm{V}$, producing the four voltage levels on each side of the bridge of $V_{\mathrm{dc}}$, $\frac{2}{3} V_{\mathrm{dc}}$, $\frac{1}{3} V_{\mathrm{dc}}$, and $0$. The output voltage can then have the seven levels

$$
v_{o} \in\left\{V_{\mathrm{dc}}, \frac{2}{3} V_{\mathrm{dc}}, \frac{1}{3} V_{\mathrm{dc}}, 0,-\frac{1}{3} V_{\mathrm{dc}},-\frac{2}{3} V_{\mathrm{dc}},-V_{\mathrm{dc}}\right\}
$$   [Eq. 08033]

# PWM

## PULSE-WIDTH-MODULATED OUTPUT

Pulse-width modulation (PWM) provides a way to decrease the total harmonic distortion of load current. A PWM inverter output, with some filtering, can generally meet THD requirements more easily than the square wave switching scheme. The unfiltered PWM output will have a relatively high THD, but the harmonics will be at much higher frequencies than for a square wave, making filtering easier.

In PWM, the amplitude of the output voltage can be controlled with the modulating waveforms. _Reduced filter requirements to decrease harmonics and the control of the output voltage amplitude are two distinct advantages of $P W M$_. Disadvantages include more complex control circuits for the switches and increased losses due to more frequent switching.

Control of the switches for sinusoidal PWM output requires (1) a reference signal, sometimes called a modulating or control signal, which is a sinusoid in this case and (2) a carrier signal, which is a triangular wave that controls the switching frequency. Bipolar and unipolar switching schemes are discussed next.

__Bipolar Switching__
Figure 08017 illustrates the principle of sinusoidal bipolar pulse-width modulation. Figure 08017a shows a sinusoidal reference signal and a triangular carrier signal. When the instantaneous value of the sine reference is larger than the triangular carrier, the output is at $+V_{\mathrm{dc}}$, and when the reference is less than the carrier, the output is at $-V_{\mathrm{dc}}$

$$
\begin{array}{ll}
v_{o}=+V_{\mathrm{dc}} & \text { for } v_{\text {sine }}>v_{\mathrm{tri}} \\
v_{o}=-V_{\mathrm{dc}} & \text { for } v_{\text {sine }}<v_{\mathrm{tri}}
\end{array}
$$   [Eq. 08034]

This version of PWM is bipolar because the output alternates between plus and minus the dc supply voltage.

The switching scheme that will implement bipolar switching using the fullbridge inverter of Fig. 08001 is determined by comparing the instantaneous reference and carrier signals:

$$
S_{1} \quad \text {and} \quad S_{2} \quad \text {are on when} \quad v_{\sin c}>v_{\text {tri }} \quad\left(v_{0}=+V_{d c}\right)
$$

$$
S_{3} \quad \text {and} \quad S_{4} \quad \text {are on when} \quad v_{\operatorname{sine}}<v_{\text {tri. }}\left(v_{\mathrm{o}}=-V_{\mathrm{dc}}\right)
$$

In a unipolar switching scheme for pulse-width modulation, the output is switched either from high to zero or from low to zero, rather than between high and low as in bipolar switching. One unipolar switching scheme has switch controls in Fig. 08001 as follows:

## PWM Definitions and Considerations

At this point, some definitions and considerations relevant when using PWM
should be stated.

##### 1. Frequency modulation ratio $m_{f}$. 
The Fourier series of the PWM output voltage has a fundamental frequency which is the same as the reference signal. Harmonic frequencies exist at and around multiples of the switching frequency. The magnitudes of some harmonics are quite large, sometimes larger than the fundamental. However, because these harmonics are located at high frequencies, a simple low-pass filter can be quite effective in removing them. Details of the harmonics in PWM are given in the next section. The frequency modulation ratio $m_{f}$ is defined as the ratio of the frequencies of the carrier and reference signals,

$$
m_{f}=\frac{f_{\text {carricr }}}{f_{\text {reference }}}=\frac{f_{\text {tri }}}{f_{\text {sinc }}}
$$   [Eq. 08035]

Increasing the carrier frequency (increasing $m_{f}$) increases the frequencies at which the harmonics occur. A disadvantage of high switching frequencies is higher losses in the switches used to implement the inverter.

##### 2. Amplitude modulation ratio $m_{a^{*}}$. 
The amplitude modulation ratio $m_{a^{*}}$ is defined as the ratio of the amplitudes of the reference and carrier signals:

$$
m_{a}=\frac{V_{m, \text { reference }}}{V_{m, \text { carrier }}}=\frac{V_{m, \text { sine }}}{V_{m, \text { tri }}}
$$   [Eq. 08036]

If $m_{a} \leq 1$, the amplitude of the fundamental frequency of the output
voltage $V_{1}$ is linearly proportional to $m_{a}$. That is,

$$
V_{1}=m_{a} V_{\mathrm{dc}}
$$   [Eq. 08037]

The amplitude of the fundamental frequency of the PWM output is thus
controlled by $m_{a^{*}}$. This is significant in the case of an unregulated dc supply voltage because the value of $m_{a^{*}}$ can be adjusted to compensate for variations in the dc supply voltage, producing a constant-amplitude output. Alternatively, $m_{a^{*}}$ can be varied to change the amplitude of the output. If ma is greater than 1, the amplitude of the output increases with $m_{a^{*}}$, but not linearly.

##### 3. Switches. 
The switches in the full-bridge circuit must be capable of carrying current in either direction for pulse-width modulation just as they did for square wave operation. Feedback diodes across the switching devices are necessary, as was done in the inverter in Fig. 08003a. Another consequence of real switches is that they do not turn on or off instantly. Therefore, it is necessary to allow for switching times in the control of the switches just as it was for the square-wave inverter.

##### 4. Reference voltage. 
The sinusoidal reference voltage must be generated within the control circuit of the inverter or taken from an outside reference. It may seem as though the function of the inverter bridge is unnecessary because a sinusoidal voltage must be present before the bridge can operate to produce a sinusoidal output. However, there is very little power required from the reference signal [it could be implemented with a simple oscillator topology]. The power supplied to the load is provided by the dc power source, and this is the intended purpose of the inverter. The reference signal is not restricted to a sinusoid, and other waveshapes can function as the reference signal.

## PWM HARMONICS

__Bipolar Switching__

The Fourier series of the bipolar PWM output illustrated in Fig. 08017 is determined by examining each pulse. The triangular waveform is synchronized to the reference as shown in Fig 08017a, and $m_{f}$ is chosen to be an odd integer. The PWM output then exhibits odd symmetry, and the Fourier series can then be expressed

$$
v_{o}(t)=\sum_{n=1}^{\infty} V_{n} \sin \left(n \omega_{0} t\right)
$$   [Eq. 08038]

For the $k$ th pulse of the PWM output in Fig. 08020, the Fourier coefficient is

$$
\begin{aligned}
V_{n k}=& \frac{2}{\pi} \int_{0}^{T} v(t) \sin \left(n \omega_{0} t\right) d\left(\omega_{0} t\right) \\
=& \frac{2}{\pi}\left[\int_{\alpha_{k}}^{\alpha_{k}+\delta_{k}} V_{d c} \sin \left(n \omega_{0} t\right) d\left(\omega_{0} t\right)+\int_{\alpha_{k}+\delta_{k}}^{\alpha_{k+1}}-\left(V_{\mathrm{dc}}\right) \sin \left(n \omega_{0} t\right)\left(d\left(\omega_{0} t\right)\right]\right.
\end{aligned}
$$

## CLASS D AUDIO AMPLIFIERS

The reference signal for the PWM control circuit can be an audio signal, and the full-bridge circuit could be used as a PWM audio amplifier. A PWM audio amplifier is referred to as a class $D$ amplifier. The triangular wave carrier signal for this application is typically $250 \mathrm{kHz}$ to provide adequate sampling, and the PWM waveform is low-pass filtered to recover the audio signal and deliver power to a speaker. The spectrum of the PWM output signal is dynamic in this case.

Class D amplifiers are much more efficient than other types of audio power amplifiers. The class AB amplifier, the traditional circuit for audio applications, has a maximum theoretical efficiency of 78.5 percent for a sine wave of maximum undistorted output. In practice, with real audio signals, class $\mathrm{AB}$ efficiency is much lower, on the order of 20 percent. The theoretical efficiency of the class $D$ amplifier is 100 percent because the transistors are used as switches. Because transistor switching and filtering are imperfect, practical class $\mathrm{D}$ amplifiers are about 75 percent efficient.

Class D audio amplifiers are becoming more prevalent in consumer electronics applications where greater efficiency results in reduced size and increased battery life. In high-power applications such as at rock concerts, class D amplifiers are used to reduce the size of the amplifier and for reduced heat requirements in the equipment.

## THREE-PHASE INVERTERS

__The Six-Step Inverter__
Figure 08028a shows a circuit that produces a three-phase ac output from a dc input. A major application of this circuit is speed control of induction motors, where the output frequency is varied. The switches are closed and opened in the sequence shown in Fig. 08028b.

Each switch has a duty ratio of 50 percent (not allowing for blanking time), and a switching action takes place every $T / 6$ time interval, or $60^{\circ}$ angle interval. Note that switches $S_{1}$ and $S_{4}$ close and open opposite of each other, as do switch pairs $\left(S_{2}, S_{5}\right)$ and $\left(S_{3}, S_{6}\right)$. As with the single-phase inverter, these switch pairs must coordinate so they are not closed at the same time, which would result in a short circuit across the source. With this scheme, the instantaneous voltages $v_{A 0}$ $v_{B 0},$ and $v_{\alpha 0}$ are $+V_{\mathrm{dc}}$ or zero, and line-to-line output voltages $v_{A B}, v_{B C},$ and $v_{C A}$ are $+V_{\mathrm{dc}}, 0,$ or $-V_{\mathrm{dc}} .$ The switching sequence in Fig. 08028b produces the output voltages shown in Fig. 08028c

The three-phase load connected to this output voltage may be connected in delta or ungrounded neutral wye. For a wye-connected load, which is the more common load connection, the voltage across each phase of the load is a line-toneutral voltage, shown in Fig. 08028d. Because of the six steps in the output waveforms for the line-to-neutral voltage resulting from the six switching transitions per period, this circuit with this switching scheme is called a six-step inverter. The Fourier series for the output voltage has a fundamental frequency equal to the switching frequency. Harmonic frequencies are of order $6 k \pm 1$ for $k=1$ $2, \ldots(n=5,7,11,13 \ldots) .$ The third harmonic and multiples of the third do not exist, and even harmonics do not exist. For an input voltage of $V_{\mathrm{dc}}$, the output for an ungrounded wye-connected load has the following Fourier coefficients:

$$
\begin{array}{l}
V_{n, L-L}=\left|\frac{4 V_{\mathrm{dc}}}{n \pi} \cos \left(n \frac{\pi}{6}\right)\right| \\
V_{n, L-N}=\left|\frac{2 V_{\mathrm{dc}}}{3 n \pi}\left[2+\cos \left(n \frac{\pi}{3}\right)-\cos \left(n \frac{2 \pi}{3}\right)\right]\right| \quad n=1,5,7,11,13, \ldots
\end{array}
$$   [Eq. 08042]

The THD of both the line-to-line and line-to-neutral voltages can be shown to be 31 percent from [Eq. 08017]. The THD of the currents in load-dependent 15 are smaller for an $R L$ load. An example of the line-to-neutral voltage and line current for an $R L$ wye-connected load is shown in Fig. 08028e.

The output frequency can be controlled by changing the switching frequency. The magnitude of the output voltage depends on the value of the dc supply voltage. To control the output voltage of the six-step inverter, the de input voltage must be adjusted.

__PWM Three-Phase Inverters__
Pulse-width modulation can be used for three-phase inverters as well as for single-phase inverters. The advantages of PWM switching are the same as for the single-phase case: reduced filter requirements for harmonic reduction and the controllability of the amplitude of the fundamental frequency.

PWM switching for the three-phase inverter is similar to that of the singlephase inverter. Basically, each switch is controlled by comparing a sinusoidal reference wave with a triangular carrier wave. The fundamental frequency of the output is the same as that of the reference wave, and the amplitude of the output is determined by the relative amplitudes of the reference and carrier waves.

As in the case of the six-step three-phase inverter, switches in Fig. 08028a are controlled in pairs $\left(S_{1},S_{4}\right)$, $\left(S_{2},S_{5}\right)$, and $\left(S_{3},S_{6}\right)$. When one switch in a pair is closed, the other is open. Each pair of switches requires a separate sinusoidal reference

wave. The three reference sinusoids are $120^{\circ}$ apart to produce a balanced threephase output. Figure 08029a shows a triangular carrier and the three reference waves. Switch controls are such that

$S_{1}$ is on when $v_{a}>v_{\text {tri }}$ 
$S_{2}$ is on when $v_{c}>v_{\text {tri }}$ 
$S_{3}$ is on when $v_{b}>v_{\text {tri }}$ 
$S_{4}$ is on when $v_{a}<v_{\mathrm{tri}}$
$S_{5}$ is on when $v_{c}<v_{\text {tri }}$ 
$S_{6}$ is on when $v_{b}<v_{\text {tri }}$

## INDUCTION MOTOR SPEED CONTROL

The speed of an induction motor can be controlled by adjusting the frequency of the applied voltage. The synchronous speed $\omega_{s}$ of an induction motor is related to the number of poles p and the applied electrical frequency $\omega$ by

$$
\omega_{s}=\frac{2 \omega}{p}
$$   [Eq. 08045]

Slip $s$, is defined in terms of the rotor speed $\omega_{r}$

$$
s=\frac{\omega_{s}-\omega_{r}}{\omega_{s}}
$$   [Eq. 08046]

and torque is proportional to slip. If the applied electrical frequency is changed, the motor speed will change proportionally. However, if the applied voltage is held constant when the frequency is lowered, the magnetic flux in the air gap will increase to the point of saturation. It is desirable to keep the air-gap flux constant and equal to its rated value. This is accomplished by varying the applied voltage proportionally with frequency. The ratio of applied voltage to applied frequency should be constant.

$$
\frac{V}{f}=\mathrm{constant}
$$   [Eq. 08047]

The term volts/hertz control is often used for this situation. The induction motor torque-speed curves of Fig. 08033 are for different frequencies and constant volts/hertz.

The six-step inverter can be used for this application if the dc input is adjustable. In the configuration of Fig. 08034, an adjustable dc voltage is produced from a controlled rectifier, and an inverter produces an ac voltage at the desired frequency. If the dc source is not controllable, a de-de converter may be inserted between the dc source and the inverter.

## SUMMARY

* The full- or half-bridge converters can be used to synthesize an ac output from a de input. 

* A simple switching scheme produces a square wave voltage output, which has a Fourier series that contains the odd harmonic frequencies of amplitudes

$$
V_{n}=\frac{4 V_{\mathrm{dc}}}{n \pi}
$$

* Amplitude and harmonic control can be implemented by allowing a zero-voltage interval of angle $\alpha$ at each end of a pulse, resulting in Fourier coefficients

$$
V_{n}=\left(\frac{4 V_{\mathrm{dc}}}{n \pi}\right) \cos (n \alpha)
$$

* Multilevel inverters use more than one dc voltage source or split a single voltage source with a capacitor voltage divider to produce multiple voltage levels on the output of an inverter. Pulse-width modulation (PWM) provides amplitude control of the fundamental output frequency. Although the harmonics have large amplitudes, they occur at high frequencies and are filtered easily. 

* Class D audio amplifiers use PWM techniques for high efficiency. The six-step inverter is the basic switching scheme for producing a three-phase ac output from a dc source. 

* A PWM switching scheme can be used with a three-phase inverter to reduce the THD of the load current with modest filtering. 

* Speed control of induction motors is a primary application of three-phase inverters.
